1/ Tìm phần tử nhỏ nhất

```python
import sys
sys.setrecursionlimit(100000)

class Node:
    def __init__(self, data = 0, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right

    def add(self, x):
        if (x < self.data):
            if (self.left):
                self.left.add(x)
            else:
                self.left = Node(x);
        elif (self.data < x):
            if (self.right):
                self.right.add(x)
            else:
                self.right = Node(x)

    def min(self):
        if (self.left):
            return self.left.min()
        else:
            return self.data

class BST:
    def __init__(self):
        self.root = None

    def add(self, data = 0):
        if (self.root is None):
            self.root =  Node(data)
        else:
            self.root.add(data)

    def min(self):
        return self.root.min()


T = BST()

n = int(input())
a = map(int, input().split())
for x in a:
    T.add(x)

print(T.min())
```

```python
import sys
sys.setrecursionlimit(100000)

class Node:
    def __init__(self, data = 0, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right

    def add(self, x):
        if (x < self.data):
            if (self.left):
                self.left.add(x)
            else:
                self.left = Node(x);
        elif (self.data < x):
            if (self.right):
                self.right.add(x)
            else:
                self.right = Node(x)

    def isLeaf(self):
        return self.left is None and self.right is None

    def sumLeaf(self):
        sum = 0
        if (self.right):
            sum += self.right.sumLeaf()
        if (self.left):
            sum += self.left.sumLeaf()
        if self.isLeaf():
            return self.data
        return sum

class BST:
    def __init__(self):
        self.root = None

    def add(self, data=0):
        if (self.root == None):
            self.root = Node(data)
        else:
            self.root.add(data)

    def sumLeaf(self):
        return self.root.sumLeaf()

T = BST()

n = int(input())
a = map(int, input().split())
for x in a:
    T.add(x)

print(T.sumLeaf())
```

8/ Tổng giá trị

```python
import sys
sys.setrecursionlimit(100000)

class Node:
    def __init__(self, data = 0, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right

    def add(self, x):
        if (x < self.data):
            if (self.left):
                self.left.add(x)
            else:
                self.left = Node(x)
        elif (self.data < x):
            if (self.right):
                self.right.add(x)
            else:
                self.right = Node(x)

    def sum_key_smaller(self, X):
        sum = 0
        if self.data < X:
            sum += self.data

        if self.left:
            sum += self.left.sum_key_smaller(X)
        if self.right:
            sum += self.right.sum_key_smaller(X)
        return sum

class BST:
    def __init__(self):
        self.root = None

    def add(self, data = 0):
        if (self.root == None):
            self.root =  Node(data)
        else:
            self.root.add(data)

    def sum_key_smaller(self, X):
        return self.root.sum_key_smaller(X)

T = BST()

n, X = map(int, input().split())
a = map(int, input().split())
for x in a:
    T.add(x)

print(T.sum_key_smaller(X))
```
