1/ Merge

```python
def merge(arr1, arr2, k):
    i = len(arr1) - 1
    j = len(arr2) - 1
    cnt = 0
    while (i >= 0 and j >= 0):
        if arr1[i] < arr2[j]:
            temp = arr1[i]
            i -= 1
        else:
            temp = arr2[j]
            j -= 1
        if cnt == k:
            return temp
        cnt += 1
    while i >= 0:
        if cnt == k:
            return arr1[i]
        i -= 1
        cnt += 1
    while j >= 0:
        if cnt == k:
            return arr2[j]
        j -= 1
        cnt += 1

def main():
    m, n, k = map(int, input().split())
    arr1 = list(map(int, input().split()))
    arr2 = list(map(int, input().split()))
    print(merge(arr1, arr2, k))

if __name__ == "__main__":
    main()
```

2/ Kiểm tra chu trình Euler

```python
def main():  
	n = int(input())  
	graph = []  
	for _ in range(n):  
		vertice = list(map(int, input().split()))  
		graph.append(vertice)  
  
	flag = True  
	for i in range(n):  
		degree = 0  
		for j in range(n):  
			degree += graph[i][j]  
		if degree % 2 != 0:  
			flag = False  
  
	if flag:  
		print("YES")  
	else:  
		print("NO")  

  
if __name__ == "__main__":  
	main()
```

3/ Biểu thức

```python
a = int(input())
b = int(input())
c = int(input())

res = [0 for i in range(6)]
res[0] = a + b + c
res[1] = a * b * c
res[2] = a + b * c
res[3] = a * b + c
res[4] = (a + b) * c
res[5] = a * (b + c)

ans = 0
for x in res:
    if x > ans:
        ans = x
print(ans)
```

4/ Ngã tư đường

```python
a = []
for i in range(4):
    a.append(list(map(int, input().split())))

crash = False

for i in range(4):
    if a[i][3] == 1:
        if a[(i + 1) % 4][0] or a[(i + 2) % 4][1] or a[(i - 1) % 4][2] or sum(a[i]) > 1:
            crash = True
            break

if crash:
    print("YES")
else:
    print("NO")
```

5/ Mật mã

```python
s = input()

t = ""
for i in range(len(s) - 1, -1, -1):
    if (i % 2 != 0):
        t += s[i]

j = 0
for i in range(len(s)):
    if (i % 2 != 0):
        print(t[j], end = "")
        j += 1
    else:
        print(s[i], end = "")
```
