6/ Tính chu vi, diện tích hình tròn

Nhập vào bán kính của một hình tròn. Tính chu vi và diện tích hình tròn.

Biết rằng:

Chu vi hình tròn = 2 * bán kính * PI

Diện tích hình tròn = bán kính * bán kính * PI

Quy ước: PI = 3.14

Dữ liệu nhập

Một dòng duy nhất gồm một số thực dương không quá 2000 là bán kính của hình tròn.

Dữ liệu xuất

Gồm hai dòng.

Dòng đầu tiên là chu vi của hình tròn.

Dòng thứ hai là diện tích của hình tròn.

Lưu ý: Kết quả phải được in ra 2 chữ số sau phần thập phân.

Ví dụ

5.0

```
radius = float(input())
PI = 3.14
P = 2 * radius * PI
S = radius ** 2 * PI
print('{0:.2f}'.format(P))
print('{0:.2f}'.format(S))
```

7/ Tính chu vi, diện tích tam giác
Viết chương trình nhập vào ba cạnh aa, bb, cc của tam giác. Tính chu vi, diện tích tam giác.

Biết rằng:

Chu vi tam giác: $P = a + b + c$
Diện tích tam giác: $S = \sqrt[2]{p * (p - a) * (p - b) * (p - c)}$

Với pp là nửa chu vi tam giác.

Dữ liệu nhập

Một dòng duy nhất gồm ba số thực aa, bb, cc (cách nhau bởi khoảng trắng) là ba cạnh của tam giác. Các cạnh có giá trị không quá 1500 và luôn đảm bảo tạo thành một tam giác.

Dữ liệu xuất

Một dòng duy nhất gồm hai số thực PP và SS cách nhau bởi một khoảng trắng. Trong đó PP là chu vi của tam giác, SS là diện tích của tam giác.

Lưu ý: Kết quả phải được in ra 2 chữ số sau phần thập phân

Ví dụ

4 5 3


```
a, b, c = map(float, input().split())
P = a + b + c
p = (a + b + c) / 2
S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
print('{0:.2f} {1:.2f}'.format(P, S))
```

8/ Tổng lương

Bạn là một kế toán của một công ty, trách nhiệm của bạn là tính lương cho các nhân viên trong công ty.

Lương của một nhân viên được tính theo công thức: Lương = Lương cơ bản * Hệ số lương + Phụ cấp.

Hãy viết chương trình để có thể tính lương của một nhân viên nhanh chóng.

Dữ liệu nhập

Dòng thứ nhất gồm một số nguyên dương $x (0 < x < 10^{8}) $ là lương cơ bản.

Dòng thứ hai gồm một số thực $a (0.0 < a < 10.0)$ là hệ số lương.

Dòng thứ ba gồm một số nguyên dương $t (0 \le t < 10^{7} $ là phụ cấp.

Dữ liệu xuất

Gồm một số, cho biết tổng lương của nhân viên trên.

Lưu ý: Kết quả phải được in ra 2 chữ số sau phần thập phân.

Ví dụ

1390000

2.34

600000

```
x = int(input())
a = float(input())
t = int(input())

salary = x * a + t

print("{:.2f}".format(salary))
```
