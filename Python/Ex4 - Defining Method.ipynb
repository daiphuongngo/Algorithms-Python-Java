1/Tổng số nguyên tố

Hãy tính tổng các số nguyên tố nhỏ hơn $N$.

Dữ liệu nhập

Một dòng duy nhất là số nguyên dương $N (N \leq 1000)$

Dữ liệu xuất

In ra tổng các số nguyên tố nhỏ hơn $N$.

Hướng dẫn giải:

Tạo hàm kiểm tra một số có phải số nguyên tố hay không: Hàm đầu vào số nguyên dương n. Đầu ra giá trị boolean ý nghĩa rằng số n có phải số nguyên tố hay không.

Xử lý:

Nếu n < 2 thì kết luận không phải số nguyên tố.

Nếu n chia hết giá trị nào từ 2 đến căn bậc hai n thì n không phải số nguyên tố.
Tạo hàm tính tổng các số nguyên tố nhỏ hơn n.

Hàm đầu vào là số nguyên dương n.

Tạo một vòng lặp chạy từ 1 đến n - 1, với mỗi lần lặp nếu nó là số nguyên tố thì cộng số đó vào tổng.

Gọi hàm tính tổng để in kết quả.

```
def isPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
sum = 0

for i in range(1, n):
    if isPrime(i) == True:
        sum += i
print(sum)
```

2/ Tính tổng

Viết chương trình tính tổng S theo công thức.

$S = 1^{2} + 2^{2} + 3^{2} + ... + n^{2}$ với $1 \leq n $

Dữ liệu nhập

Gồm một dòng duy nhất chứa số nguyên dương $n (n \leq 1000)$.

Dữ liệu xuất

Gồm một dòng duy nhất in ra kết quả bài toán.

```
def solve(n):
    ans = 0
    for i in range(1, n+1):
        ans += i**2
    return ans

n = int(input())
result = solve(n)
print(result)
```

3/ Rút gọn phân số

Cho phân số $\frac{a}{b}$ với $a$ và $b$ là hai số nguyên dương. Hãy rút gọn phân số đã cho.

Phân số $\frac{c}{d}$ gọi là phân số rút gọn của $\frac{a}{b}$ khi:

$c = \frac{a}{GCD(a,b)}$

$d = \frac{b}{GCD(a,b)}$

Dữ liệu nhập

Gồm một dòng duy nhất chứa 2 số nguyên dương $a, b$ cách nhau bởi một khoảng trắng $(a, b \leq 10000)$.

Dữ liệu xuất

In ra trên một dòng duy nhất hai số $c$ và $d$ là tử và mẫu của phân số tối giản, cách nhau bởi một khoảng trắng.

Hướng dẫn giải:

Khái niệm GCD(a,b) là ước chung lớn nhất của a, b.

Ý tưởng sử dụng thuật chia Euclidean:

Ước chung lớn nhất của a, b chính là bằng ước chung lớn nhất của b và a mod b.
Giải thích gọi c là ước chung lớn nhất của a và b vậy có thể viết a = a’.c, b = b’.c, a cũng có thể viết thành a = b.x + y với y = a mod b. Do b.x là bội của c do b là bội của x nên chỉ cần y là bội của c là c sẽ trở thành ước của a.

Hướng dẫn giải

Sử dụng thuật chia Euclidean.

Nếu a còn khác 0 ước chung lớn nhất của a, b chính là ước chung lớn nhất của b và a mod b.

Nên ta sử dụng vòng lặp cập nhật a, b thành b, a%b đến khi nào a nhận giá trị bằng 0.

Gọi hàm tìm ước chung để tính kết quả.

Từ giá trị GCD vừa tìm được áp dụng tính c, d theo công thức đề cho.

Độ phức tạp: O(log(min(a, b))) với a, b là tử và mẫu của phân số cần tối giản.

```
def GCD(a, b):
	while b != 0:
		r = a % b
		a = b
		b = r
	return a

def reduceFraction(a, b):
	gcd = GCD(a, b)
	a /= gcd
	b /= gcd
	return a, b

numerator, denominator = map(int, input().split())
print("%d %d" % reduceFraction(numerator, denominator))
```

4/ Đếm số lượng chữ số

Cho một số nguyên dương $N$, hãy cho biết số $N$ có bao nhiêu chữ số.

Dữ liệu nhập

Một số nguyên dương $N (N \leq 10^{9})$

Dữ liệu xuất

In ra số lượng chữ số của $N$.

```
def countDigit(n):
    ans = 0
    while n > 0:
        ans += 1
        n = n // 10
    return ans

n = int(input())
print(countDigit(n))
```

5/ Chữ số lớn nhất

Cho một số nguyên dương $N$, hãy cho biết chữ số lớn nhất của số $N$.

Dữ liệu nhập

Một số nguyên dương $N (N \leq 10^9)$

Dữ liệu xuất

In ra chữ số lớn nhất của số $N$.

Hướng dẫn giải

Tạo hàm tìm chữ số lớn nhất trong số nn:

Đầu vào hàm là một số nguyên nn.

Giá trị đầu ra là chữ số lớn nhất.

Ban đầu gán biến biến kết quả với giá trị -1.

Ta sẽ tách từng chữ số ra từ số n bằng vòng lặp không biết trước:

Khi n > 0 thì kiểm tra, nếu số dư của phép chia n cho 10 lớn hơn biến kết quả thì cập nhật lại biến kết quả bằng số dư của phép chia của nn cho 10.

Sau đó chia lấy phần nguyên của n cho 10.

Gọi hàm tìm chữ số lớn nhất để in kết quả.

```
def maxDigit(n):
    ans = -1
    while n > 0:
        if ans < n % 10:
            ans = n % 10
        n = n // 10
    return ans

n = int(input())
print(maxDigit(n))
```

7/ Tìm số nguyên tố gần $N$ nhất

Cho một số nguyên dương $n$, hãy tìm số nguyên tố $m$ gần nhất với $n$.

Nếu $n$ là số nguyên tố thì $m = n$.

Giả sử có ba số $a < b < c$, số $a$ được nói gần với $b$ hơn so với $c$ khi: $b - a < c - b$

Nếu có hai số cùng gần với $n$ thì chọn số nhỏ hơn.

Dữ liệu nhập

Một số $n$ duy nhất $(2 \leq n \leq 10^4)$

Dữ liệu xuất

In ra số nguyên tố gần với $n$.

Giải thích ví dụ

Ví dụ 1: Do 5 chính là số nguyên tố nên kết quả là 5.

Ví dụ 2: Xung quanh 4 ta có hai số nguyên tố là 3, 5 (cùng cách một đơn vị), ta sẽ chọn 3.

Ví dụ 3: Xung quanh 10 ta có hai số nguyên tố là 7 và 11, nhưng 11 ở gần hơn nên ta chọn 11.

Hướng dẫn giải

Tạo hàm kiểm tra một số có phải số nguyên tố hay không:

Hàm đầu vào số nguyên dương n.

Đầu ra giá trị boolean ý nghĩa rằng số n có phải số nguyên tố hay không.

Xử lý

Nếu n < 2 thì kết luận không phải số nguyên tố.

Nếu n chia hết cho giá trị nào từ 2 đến căn bậc hai của n thì n không phải số nguyên tố.

Dùng hàm nhập dữ liệu để đọc số nguyên n.

Tạo 2 biến mới có giá trị bằng n.


Sử dụng vòng lặp không biết trước để tìm số nguyên tố nhỏ hơn, và số nguyên tố lớn hơn gần n nhất.

Vòng lặp thứ nhất, trừ biến thứ nhất cho đến khi nó là số nguyên tố khi đó số đó là số nguyên tố gần nhất nhỏ hơn n.

Vòng lặp thứ hai, cộng biến thứ hai cho đến khi nó là số nguyên tố khi đó số đó là số nguyên tố gần nhất lớn hơn n.

Cuối cùng kiểm tra số nguyên tố lớn hơn và nhỏ hơn n xem số nào gần n hơn bằng cách kiếm tra 2 hiệu: Nếu n trừ biến thứ nhất nhỏ hơn biến thứ 2 trừ n thì biến thứ nhất là kết quả cần tìm, ngược lại biến thứ 2 là kết quả cần tìm.

```
def isPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())

n1 = n
n2 = n

while isPrime(n1) == False:
    n1 -= 1

while isPrime(n2) == False:
    n2 += 1

if n - n1 <= n2 - n:
    print(n1)
else:
    print(n2)
```

8/ Kiểm tra số đối xứng

Cho số nguyên dương $n$, kiểm tra số $n$ có phải là số đối xứng hay không. Biết rằng số đối xứng là số mà đọc từ trái sang phải (không bao gồm các chữ số 0 vô nghĩa ở đầu) hay từ phải sang trái đều như nhau.

Ví dụ số 12321 là số đối xứng, 1221 là đối xứng, 5656 không phải số đối xứng, 100 không phải là số đối xứng.

Dữ liệu nhập

Một dòng duy nhất là số nguyên dương $n (n \leq 1.000.000)$.

Dữ liệu xuất

Nếu nn là số đối xứng in ra "YES", ngược lại in ra "NO".

Không in dấu "".

```
# Method 1:
if __name__ == '__main__':
    A = input()
    B = A[::-1]
    if B == A:
        print('YES')
    else:
        print('NO')
```

Hướng dẫn giải

Để ý rằng một số là số đối xứng khi và chỉ khi số đảo ngược của nó bằng chính nó. Ví dụ:

n = 12321 thì số đảo ngược của nó là 12321 $\Rightarrow$ n là số đối xứng.

n = 1212 thì số đảo ngược của nó là 2121 $\Rightarrow$ n không là số đối xứng.

Với ý tưởng này, ta sẽ xây dựng hàm reverse(n) để có số đảo ngược và so sánh nó có bằng với số nn ban đầu không. Nếu có in ra "YES", ngược lại in ra "NO".

```
# Method 2:
def reverse(n):
    ans = 0
    while n != 0:
        ans = ans * 10 + n % 10
        n = n // 10
    return ans

def isSymmetry(n):
    return n == reverse(n)

n = int(input())
if isSymmetry(n) == True:
    print("YES")
else:
    print("NO")
```
