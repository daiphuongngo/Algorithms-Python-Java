1/ Cấu trúc phân số

Cho hai phân số, hãy viết chương trình tính tổng và rút gọn hai phân số đó.

Dữ liệu nhập

Dòng thứ nhất là thông tin tử số và mẫu số của phân số thứ nhất.

Dòng thứ hai là thông tin tử số và mẫu số của phân số thứ hai.

Các số trong dữ liệu đều thuộc phạm vi số nguyên dương không quá 1.000.

Dữ liệu xuất

Một dòng duy nhất là hai số nguyên aa và bb cách nhau bởi khoảng trắng, với a, ba,b là tử và mẫu của tổng hai phân số từ dữ liệu vào.

Ví dụ

Giải thích ví dụ

Phân số đầu là $\frac{1}{3}$

​​  và phân số thứ hai là $\frac{1}{3}$

​​ . Vậy tổng 2 phân số là $\frac{2}{3}$


Hướng dẫn giải:

Cấu trúc để giải quyết bài này là phân số. Chúng ta biết một phân số sẽ đuợc định nghĩa bằng 2 thành phần là tử và mẫu. Như vậy ta đã có một cái nhìn tổng thể cho cấu trúc này, đó là lớp cấu trúc này có thể đặt tên là FractionFraction, và có 2 thuộc tính là numerator (tử số) và denominator (mẫu số).

Tiếp theo, để có thể tạo ra các đối tuợng có cấu trúc phân số, ta cần cài đặt cho lớp phân số một phuơng thức khởi tạo (hay hàm khởi tạo).

Tiếp đến, đề bài yêu cầu chúng ta cộng 2 phân số với nhau. Để thực hiện việc này một cách huớng đối tuợng, chúng ta sẽ tạo một phuơng thức cộng cho mỗi đối tuợng lớp FractionFraction.

Trong hàm này chúng ta sẽ tính tử và mẫu số của phân số tổng mới rồi trả về một phân số đuợc khởi tạo từ tử và mẫu vừa tính đó (việc tính tổng 2 phân số đuợc thực hiện bằng cách quy đồng mẫu số):

Tính tử và mẫu của phân số mới

Tạo và trả về phân số mới có tử và mẫu vừa tính

Việc cuối cùng đề bài yêu cầu là rút gọn phân số. Như vậy ta cũng sẽ tạo một phuơng thức để rút gọn cho từng đối tuợng lớp Fraction.

Thuật tóan rút gọn phân số bạn có thể xem lại ở bài L04P04 (chúng ta sẽ chia tử và mẫu cho uớc số chung lớn nhất của chúng)

Lưu ý: các khai báo phuơng thức phải nằm trong khai báo lớp thì mới gọi là phuơng thức và sử dụng đuợc trên các đối tuợng, nguợc lại nó chỉ là hàm.

```python
def gcd(a, b):
    while (b != 0):
        r = a % b
        a = b
        b = r
    return a
class PhanSo:
    def __init__(self, nu, de = 1):
        self.nu = nu
        self.de = de
    def cong(self, other):
        a = self.nu * other.de + other.nu * self.de
        b = self.de * other.de
        return PhanSo(a, b)
    def reduce(self):
        c = gcd(self.nu, self.de)
        self.nu //= c
        self.de //= c

x, y = map(int, input().split(' '))
a = PhanSo(x, y)

x, y = map(int, input().split(' '))
b = PhanSo(x, y)

c = a.cong(b)
c.reduce()

print(c.nu, c.de)
```

2/ Tính điểm xa nhất

Cho bạn một điểm M(x, y) và N điểm khác trong mặt phảng Oxy, hãy tìm những điểm xa nhất so với điểm M. Biết công thức tính khoảng cách giữa hai điểm A(x, y) và B(z, t) là:

distance(A,B) = $\sqrt{(x-z)^{2} + (y-t)^{2}}$
​​ 
Dữ liệu nhập

NN dòng tiếp theo, mỗi dòng là hai số nguyên $x_{i}$
​​  và $y_{i}$
​​ , là tung độ và hoành độ của điểm thứ i (0 $\leq$ i < N).

Các số tọa độ điểm đều là số nguyên nằm trong đoạn [-1000, 1000].

Nếu có nhiều điểm cùng đều cách xa nhất so với M, chỉ cần lấy điểm đầu tiên theo thứ tự trong dữ liệu vào.

Giải thích ví dụ

Ví dụ 1:

Hướng dẫn giải:

Chúng ta tạo lớp cấu trúc tên PointPoint để lưu một điểm trong không gian 2 chiều. Một đối tuợng PointPoint có 2 thuộc tính là x (hòanh độ) và y (tung độ).

Hàm khởi tạo cũng sẽ khởi tạo đối tuợng dựa trên hòanh độ và tung độ

Ngoài ra mỗi điểm cần có thêm phuơng thức tính khỏang cách đến một điểm khác. Đặt tên phương thức này là distance.

Khỏang cách giữa self và other sẽ đuợc tính theo công thức khỏang cách Euclide:
return $((self.x – other.x)^{2} + (self.y – other.y)^{2})^{0.5}
$
​​ 
Sau khi chuẩn bị xong các phương thức trên, ta chỉ dần duyệt một lần qua các điểm trong mảng và cập nhật khoảng cách lớn nhất từ 1 điểm tới M.

```python
class Point:  
    def __init__(self, a, b):  
        self.x = a  
        self.y = b  
  
    def dist(self, other):  
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5  
  
    def clone(self):  
        tmp = Point(self.x, self.y)  
        return tmp  
  
  
x, y = map(int, input().split())  
M = Point(x, y)  
n = int(input())  
a = []  
for i in range(n):  
    x, y = map(int, input().split())  
    a.append(Point(x, y))  
  
maxLen = 0  
ans = M.clone()  
for point in a:  
    curLen = M.dist(point)  
    if curLen > maxLen:  
        maxLen = curLen  
        ans = point  
print(ans.x, ans.y)  
```

4/ Tìm tổng diện tích các tam giác

Lần lượt cho hoành độ và tung độ 3 cạnh của 1 mảng các tam giác. Hãy tính tổng diện tích các tam giác trong mảng. Cho công thức của diện tích của 1 tam giác như sau với ss là nửa chu vi của tam giác và a, b, ca,b,c lần lượt là độ dài các cạnh.

$A = \sqrt{s(s-a)(s-b)(s-c)}$
​​ 
Dữ liệu nhập

Tung độ và hoành độ của mỗi điểm sẽ là 3 số nguyên trong đoạn [-1000, 1000]. Dữ liệu đảm bảo rằng 3 điểm của 1 tam giác sẽ không thẳng hàng.

Dữ liệu xuất

In ra 1 dòng duy nhất là tổng diện tích của các tam giác trong mảng. Kết quả chỉ in ra với hai chữ số thập phân.

Giải thích ví dụ

Ví dụ 1:

​
​​ . Vậy diện tích tam giác là 0.5.

Hướng dẫn giải:

Ở bài này chúng ta cần sử dụng ít nhất lớp cấu trúc, đó là điểm. Cấu trúc điểm ta có thể tận dụng ở bài L10P03.

Ban đầu ta khởi tạo một biến result = 0 để lưu lại tổng kết quả diện tích các tam giác

Ta sẽ duyệt qua từng cặp 3 điểm của mỗi tam giác. Sau đó, từ 3 điểm ấy ta sẽ tính được độ dài 3 cạnh của tam giác. Từ độ dài 3 cạnh của 1 tam giác, ta có thể dễ dàng tính được diện tích của một tam giác bằng công thức Heron như ở trong đề.

$p = \frac{(AB + BC + CA)}{2}$
​​ 

$Area = sqrt{(p * (p – AB) * (p – BC) * (p – CA))}$

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5

    def __str__(self):
        return '{:.2f} {:.2f}'.format(self.x, self.y)


n = int(input())
res = 0
for i in range(n):
	x, y = map(int, input().split())
	A = Point(x, y)
	x, y = map(int, input().split())
	B = Point(x, y)
	x, y = map(int, input().split())
	C = Point(x, y)
	a = B.distance(C)
	b = A.distance(C)
	c = B.distance(A)
	p = (a+b+c)/2
	area = math.sqrt(p*(p-a)*(p-b)*(p-c))
	res += area
print("%.2f" % (res))
```

Chúng ta sẽ lưu danh sách các học sinh và thực hiện tìm kiếm trên danh sách đó. Các thao tác hòan tòan tuơng tự như trên các số nguyên:

Truy vấn: với mỗi mã số truy vấn (gọi là query_id), chúng ta chỉ cần duyệt từng học sinh trong danh sách cho đến khi gặp học sinh có mã số trùng với query_id thì in điểm tóan và văn của học sinh đó ra và dừng tìm kiếm.

Ngoài ra bài này còn có thể giải bằng một cách “có tổ chức” hơn đó là sử dụng thêm một lớp School (nhà truờng), có thuộc tính là danh sách các học sinh, và việc thêm học sinh cũng như tìm kiếm học sinh sẽ trở thành các phuơng thức của lớp School này.
```python
class Student:
    def __init__(self, id, math, literature):
        self.id = id
        self.math = math
        self.literature = literature

n, q = map(int, input().split(' '))
students = []
for i in range(n):
    id, math, literature = input().split(' ')
    math = float(math)
    literature = float(literature)
    new_student = Student(id, math, literature)
    students.append(new_student)

for i in range(q):
    query_id = input()
    for student in students:
        if student.id == query_id:
            print(student.math, student.literature)
            break
```

Bài này thoạt nhìn có thể dễ nghĩ tới cách tổ chức cấu trúc là mỗi một băng giấy trong n băng giấy đuợc cho là một đối tuợng lớp Băng giấy, với thuộc tính là màu và kích thuớc. Tuy nhiên bạn sẽ thấy cách này không đuợc hay nếu đi vào cài đặt cụ thể.

Thực ra cấu trúc chúng ta cần cài đặt ở bài này đuợc mô tả ở phần đầu ra đề bài. Mục tiêu của chúng ta là thống kê mỗi loại màu có số luợng và tổng kích thuớc như thế nào. Vì vậy ta nên tạo một cấu trúc Color, với 2 thuộc tính là sum_length (tổng độ dài các băng màu có màu là color_id) và count (số luợng các băng màu có màu là color_id). Đây là thông tin thống kê của màu color_id.

Ta sẽ sử dụng một mảng colors với ý nghĩa: colors[i] sẽ là thông tin của màu i. Do ta có đến 1000 màu, vì vậy ta sẽ truy cập đến colors[1000]. Suy ra, mảng của ta phải lưu đến 1001 phần tử.

Như vậy với mỗi băng màu có mã màu p và độ dài l đuợc nhập vào thì ta chỉ cần cập nhật thông tin của colors[p], sử dụng một phuơng thức cập nhật cho phép tăng sum_length của colors[p] lên l đơn vị và count của colors[p] lên 1 đơn vị.

Số luợng màu khác nhau chính là số luợng các phần tử trong mảng colors có giá trị count > 0 (hoặc sum_length > 0).

Và cuối cùng để in dữ liệu ta chỉ cần duyệt i từ 1 đến 1000, phần tử nào có count > 0 thì màu đó có tồn tại.

Lưu ý

Đối với Java, bạn xem kĩ phần khai báo !

```python
class Color:
    def __init__(self, color_id):
        self.color_id = color_id
        self.sum_length = 0
        self.count = 0
 
    def update(self, new_length):
        self.sum_length += new_length
        self.count += 1
 
n = int(input())
 
colors = []
for i in range(1001):
    colors.append(Color(i))
 
for i in range(n):
    p, l = map(int, input().split(' '))
    colors[p].update(l)
 
different_colors = 0
for i in range(1, 1001):
    if colors[i].count > 0:
        different_colors += 1
 
print(different_colors)
 
for i in range(1, 1001):
    if colors[i].count > 0:
        print(i, colors[i].sum_length, colors[i].count)
```

Hướng dẫn giải

Bài này cấu trúc ta cần tổ chức rõ ràng là nhân viên. Đặt tên lớp cấu trúc này là Staff, mỗi đối tuợng sẽ có 3 thuộc tính như đề bài yêu cầu đó là id (mã số), name (tên) và birth_year (năm sinh).

Ban đầu chúng ta có thể lưu tất cả nhân viên vào một danh sách staffs.

Sau đó, để giải bài này ta lại sử dụng một kỹ thuật quen thuộc đó là đặt lính canh:

Đầu tiên giả sử nhân viên lớn tuổi nhất là nhân viên staffs[0]. Gán cho biến lính canh là oldest_staff = staffs[0].

Với mỗi nhân viên còn lại, ta sẽ so sánh nguời đó với nhân viên lớn tuổi nhất hiện tại. Giả sử đang xét nhân viên thứ i, thì ta sẽ chỉ cập nhật lại oldest_staff = staffs[i] khi mà:

staffs[i].birth_year < oldest_staff.birth_year

Hoặc staffs[i].birth_year == oldest_staff.birth_year và staffs[i].id < oldest_staff.id

Cuối cùng oldest_staff chính là nhân viên lớn tuổi nhất cần tìm.

```python
class Staff:
    def __init__(self, id, name, birth_year):
        self.id = id
        self.name = name
        self.birth_year = birth_year

n = int(input())

staffs = []

for i in range(n):
    id, name, birth_year = input().split(' ')
    birth_year = int(birth_year)
    staffs.append(Staff(id, name, birth_year))

oldest_staff = staffs[0]
for i in range(1, n):
    if staffs[i].birth_year < oldest_staff.birth_year or (staffs[i].birth_year == oldest_staff.birth_year and staffs[i].id < oldest_staff.id):
        oldest_staff = staffs[i]

print(oldest_staff.id, oldest_staff.name, oldest_staff.birth_year)
```

Hướng dẫn giải

Cấu trúc để giải quyết bài này là phân số.
Cấu trúc phân số tương tự ở bài một tuy nhiên ngoài các hàm đã thực hiện là hàm khởi tạo hay hàm rút gọn phân số ta cần một hàm so sánh để tìm ra phân số lớn nhất.

Giả sử có hai phân số $\frac{a}{b}$
​​  và $\frac{c}{d}$

Ta nói phân số $\frac{a}{b}$
​​  lớn hơn $\frac{c}{d}$
​​  khi $\frac{a}{b} - \frac{c}{d} > 0$
hay $\frac{ad - bc}{bd} > 0$

Do $b \times d > 0$ vì b và d đều là số nguyên dương nên chỉ cần $ad - bc > 0$.

Tương tự như trên ta tạo hàm trả về giá trị chênh lệnh:

Nếu compare(self, other) > 0 ta nói phân số self lớn hơn phân số other.

Nếu compare(self, other) < 0 ta nói phân số self nhỏ hơn phân số other.

Nếu compare(self, other) = 0 ta nói phấn số self bằng phân số other.

Tiến hành giải bài toán:

Ta đọc tất cả phân số vào một mảng phân số.

Sử dụng kĩ thuật đặt lính canh tìm phân số lớn nhất thông qua hàm compare.

Rút gọn phân số lớn nhất và in kết quả.

```python
class Fraction:
    def __init__(self, x, y):
        self.nu = x
        self.de = y
    def compare(self, other):
        return self.nu * other.de - self.de * other.nu
    def GCD(self):
        a = self.nu
        b = self.de
        while (b != 0):
            r = a % b
            a = b
            b = r
        return a
    def reduce(self):
        gcd = self.GCD()
        self.nu //= gcd
        self.de //= gcd

n = int(input())

lst = []

for i in range(n):
    x, y = map(int, input().split(' '))
    lst.append(Fraction(x, y))

ans = lst[0]
for x in lst:
    if (ans.compare(x) <= 0): # x, not lst[i]
        ans = x # x, not lst[i]

ans.reduce()
print(ans.nu, ans.de)
```
