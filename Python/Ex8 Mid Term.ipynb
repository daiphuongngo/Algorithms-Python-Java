1/ Taxi Fare

Tính tiền taxi

Viết chương trình tính tiền đi taxi dựa vào số km nhập vào (số nguyên) theo công thức sau:

Giá tiền đi taxi theo km	Giá (vnd/km)

1 km đầu tiên	15,000

Từ km thứ 2 đến thứ 5	13,500

Từ km thứ 6 trở lên	11,000

Lưu ý: Nếu đi từ km thứ 12 trở lên được giảm 10% tổng tiền.

Dữ liệu nhập

Một dòng duy nhất, chứa một số nguyên nn là số km mà bạn đã đi được $(0 < n \leq 100)$.

Dữ liệu xuất

In ra một số nguyên duy nhất là số tiền taxi bạn phải trả.

Ví dụ

input

4

output

55500

input

15

output

161100

Giải thích ví dụ

Ví dụ 1:

1km đầu được tính với giá 15.000.

Từ km thứ 2 đến thứ 4 được tính với giá 13.500.

Vậy tổng cộng là 15.000 + 13.5000 * 3 = 555.000.

Ví dụ 2:

1km đầu được tính với giá 15.000.

Từ km thứ 2 đến thứ 5 được tính với giá 13.500.

Từ km thứ 6 đến 15 được tính với giá 11.000.

Vậy tổng cộng là 15.000 + 13.500 * 4 + 11.000 * 10 = 179.000. Nhưng theo chính sách của hãng taxi thì họ được giảm 10%, vậy số tiền thực tế họ phải trả là 161.100.

```
n = int(input())  
vnd = 0  
if n <= 1:  
    vnd = 15000  
elif n <= 5:  
    vnd = 15000 + (n - 1) * 13500  
else:  
    vnd = 15000 + 4 * 13500 + (n - 5) * 11000  
   
if n >= 12:  
    vnd = vnd * 0.9  
   
print(int(vnd))  
```

```
# Method 1: (mine)
def staircase(n):
    size = n-1
    step = 1
    for i in range(n):
        string = ' '*(size) + '*'*(step)
        step += 2
        print(string)
        size -= 1
if __name__ == '__main__':
    n = int(input())

    staircase(n)
```

4/ Đếm Nam châm - Count Magnet

Nhà bác học điên Mike rất thích tiêu khiển bằng trò chơi domino. Nhưng vì không tìm được quân domino nào, ông ta đành dùng các thỏi nam châm thay thế. Mỗi nam châm có hai cực: cực dương (+) và cực âm (-). Khi đưa hai nam châm lại gần nhau nếu hai cực có cùng bản chất (cùng dương hoặc cùng âm) sẽ đẩy nhau, khác bản chất sẽ hút nhau.

Ông ta bắt đầu xếp những thỏi nam châm liên tiếp lên bàn theo chiều ngang. Tùy theo cách sắp xếp của Mike mà các thỏi nam châm hút hay đẩy nhau tạo thành những nhóm nam châm riêng biệt.

Nhiệm vụ của bạn là đếm xem có bao nhiêu nhóm nam châm được tạo thành.

Hình dưới đây minh họa một cách sắp xếp nam châm: 

Dữ liệu nhập

Dòng đầu tiên chứa một số nguyên $n (1 \leq n \leq 100.000)$ là số lượng nam châm.

$n$ dòng tiếp theo mỗi dòng chứa một cặp kí tự “01” nếu Mike đặt nam châm theo chiều (+)(-) hoặc “10” nếu theo chiều ngược lại.

Dữ liệu xuất

Một dòng duy nhất là số nhóm nam châm được tạo thành.

Ví dụ

input

6

10

10

10

01

10

output

3

```
# Method 1:
def magnet(n, m): 
    count = 1
    for i in range(1, n): 
        if (m[i] != m[i - 1]): 
            count += 1
    return count 
  
if __name__ == "__main__": 
    n = int(input())
    m = []
    for i in range(n):
        mag = int(input())
        m.append(mag)
    print(magnet(n, m)) 
```

Hướng dẫn giải

Ta biết rằng nếu 2 cặp nam châm liên tiếp khác nhau (01 – 10) thì sẽ tạo ra thêm một nhóm nam châm.

Ví dụ: Ta có 4 cặp nam châm 01-10-10-01. Lần lượt xét từng cặp:

01: số nhóm = 1

10: so sánh với cặp kề trước $10 \neq 01 \rightarrow$ số nhóm = 2

10: so sánh với cặp kề trước $10 = 10 \rightarrow$ số nhóm = 2

01: so sánh với cặp kề trước $01 \neq 10 \rightarrow$ số nhóm = 3

Từ đó ta suy ra có ý tưởng để giải bài này như sau:

Dùng 2 biến $pre$ và $cur$ để lưu 2 cặp nam châm liên tiếp nhau. So sánh $pre$ và $cur$, nếu $pre$ khác $cur$ thì tăng số nhóm lên 1. Sau đó cập nhật lại $pre$. Cứ như thế cho đến cặp cuối cùng ta thu được kết quả là số nhóm nam châm tạo thành.

Khởi tạo: số nhóm = 1.

Nhập vào $pre$

Cho vòng lặp chạy từ $1 \rightarrow n - 1$:

Nhập vào $cur$

Nếu $pre \neq cur$:

Số nhóm + 1

Cập nhật $pre = cur$

In ra kết quả bài toán

```
# Method 2:
n = int(input())
pre = input()
count = 1
for i in range(1, n):
	cur = input()
	if cur[0] == pre[1]: # string type, define the location of character in string 
		count += 1
	pre = cur # update again (lật pin)
print(count)
```

5/ Viết tắt

Cho bạn một chuỗi là tên của một bài hát. Để dễ tìm kiếm, đề bài yêu cầu in ra tên viết tắt chữ cái đầu của bài hát đó dưới dạng chữ hoa.

Dữ liệu nhập

Một dòng duy nhất chứa tên bài hát (không quá 1000 ký tự). Đảm bảo không có khoảng trắng đầu và cuối câu, giữa 2 từ là 1 khoảng trắng, các ký tự có thể viết hoa thường tùy ý.

Dữ liệu xuất

In ra một dòng duy nhất là chuỗi ký tự viết tắt tên bài hát đó bằng chữ hoa.

Ví dụ

input

Cotton Eye Joe

output

CEJ

Hướng dẫn giải:

Các bước thực hiện:

Bước 1: Đọc chuỗi ss của input.

Bước 2: Chuyển đổi tất cả các kí tự trong ss thành chữ hoa.

Bước 3: Tạo chuỗi kết quả acronym rồi thêm kí tự đầu tiên vào chuỗi.

Bước 4: Duyệt qua chuỗi đã chuyển thành chữ hoa từ index 1 đến cuối chuỗi, tiến hành cộng kí tự $s[i]$ vào acronym nếu $s[i - 1]$ là khoảng trắng.

Bước 5: In chuỗi acronym.

Độ phức tạp: $O(len(s))$ với $len(s)$ là độ dài của chuỗi.

```
# Method 1: (mine)
def shorten(n):
    lst = []
    for i in range(len(n)):
      if n[i] != ' ': # bỏ hết space
        item = n[i][0].capitalize()  # C, E, J
        lst.append(item) # ['C', 'E', 'j']
    listToStr = ''.join(map(str, lst)) # CEJ
    print(listToStr)
if __name__ == '__main__':
    n = list(map(str, input().rstrip().split())) # ['cotton', ' ', 'Eye', ' ', 'Joe']

    shorten(n)
```
