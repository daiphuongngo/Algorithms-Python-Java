5/ Đếm số nguyên tố

Cho một mảng các số nguyên, hãy đếm xem mảng có bao nhiêu số nguyên tố.

Dữ liệu nhập

Dòng đầu tiên bao gồm một số nguyên dương nn là kích thước mảng $(n \leq 1000)$

Dòng thứ hai gồm nn số nguyên $a_i$ là các phần tử của mảng, cách nhau bởi các khoảng trắng $(|a_i| \leq 1.000.000)$

Dữ liệu xuất

Một dòng duy nhất là số lượng số nguyên tố trong mảng.

Có thể chứng minh được chỉ cần cho i chạy từ 2 đến sqrt(x) là có thể kiểm tra số đó có phải là số nguyên tố hay ko như sau:

Khi $N$ chia hết cho 1 số nào đó, thì ta có thể phân tích $N$ thành tích của 2 số

Giả sử $N = a*b với 1 < a <= b < N$

Ta có: $N = a*b >= a * a = a^2$

Hay $a^2 <= N$

Suy ra $1 < a <= sqrt(N)$

Do đó ta chỉ cần xét N có chia hết cho a chạy từ $[2, sqrt(N)]$ ko, vì nếu N chia hết cho a trong khoảng đó, thì sẽ tìm được b tương ứng nên ko cần phải xét hết từ $[2, N-1]$

```
def isPrime(x):
    if x < 2: 
        return False
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True

n = int(input())
a = list(map(int, input().split()))

count = 0
for i in range(n):
    if isPrime(a[i]):
        count += 1
print(count)
```

6/ Tỉa hoa

Henry quyết định tỉa lại hoa trong khu vườn của mình.

Khu vườn của Henry có n chậu hoa, mỗi cây trong chậu hoa có một chiều cao là $h_i$ (cm)

Theo Henry một khu vườn đẹp là khu vườn có n chậu hoa cùng chiều cao với nhau. Vì vậy ông quyết định tỉa hoa trong vườn sao cho chúng có cùng chiều cao. Nhưng mỗi lần tỉa Henry chỉ có thể tỉa được 1 chậu và đúng 1cm và mất 1 đơn vị năng lượng.

Bạn hãy cho biết năng lượng tối thiểu mà Henry có thể tạo ra khu vườn với chiều cao theo ý muốn của mình.

Dữ liệu nhập

Dòng đầu tiên chứa số nguyên dương $n (n \le 100.000)$ là số chậu hoa trong vườn.

Dòng thứ hai chứa n số nguyên dương là chiều cao của chậu hoa tương ứng $h_i$  $(h_i \le 1000, 0 \le i < n)$

Dữ liệu xuất

In ra đơn vị năng lượng nhỏ nhất là Henry phải tiêu tốn.

Hướng dẫn giải

Đọc dữ liệu:

Đọc số nn nằm trên một dòng duy nhất.
Đọc dãy gồm các phần tử là số nguyên vào mảng hh.
Có thể quan sát thấy là phương án tỉa cây tốt nhất là tỉa tất cả các cây về cùng độ cao với cây thấp nhất, bởi vì ta không thể tỉa các cây về chiều cao lớn hơn so với chiều cao của cây hiện tại, và nếu tỉa thấp hơn cây thấp nhất thì sẽ rất tốn công vô ích. Do đó thuật toán để giải quyết bài này gồm có 2 phần: tìm cây có chiều cao nhỏ nhất, và tính tổng năng lượng để tỉa tất cả các cây về cùng chiều cao này.

Tìm cây thấp nhất: ta lại sử dụng kỹ thuật đặt lính canh như ở bài 2 (tìm phần tử lớn nhất trong mảng), chỉ khác là lần này ta sẽ tìm phần tử nhỏ nhất trong mảng hh. Đầu tiên ta khởi tạo biến lính canh là minH bằng một số rất lớn (chỉ cần lớn hơn hoặc bằng giá trị lớn nhất mà $h[i]$ có thể nhận), ví dụ như 1001. Duyệt từng phần tử $h[i]$ trong mảng (sử dụng vòng lặp for, nếu h[i] "tốt" hơn minHminH (tức là $h[i] < minHh[i])$ thì lúc này ta coi $h[i]$ là $minH$ (cập nhật lại $minH = h[i]$).


Tính tổng năng lượng: với cây thứ i, ban đầu nó có chiều cao là $h[i]$, giờ ta muốn tỉa nó xuống chiều cao là minH, thì ta sẽ tỉa $h[i] - minH$ $(cm)$ chiều cao cây, như vậy năng lượng để tỉa cây thứ $i$ là $h[i] - minH$. Vậy ban đầu ta sẽ khởi tạo biến kết quả là energy = 0. Sau đó xét từng cây $h[i]$ bằng vòng lặp for, rồi cộng thêm $h[i] - minH$ vào kết quả: $energy = energy + (h[i] - minH)$.

```
n = int(input())
h = list(map(int, input().split()))
minH = 1001
for i in range(n): # luôn tìm ra đc min Height nên áp mới min Height vào existing one
    if h[i] < minH:
        minH = h[i]
energy = 0
for i in range(n):
    energy = energy + (h[i] - minH) # enery ban đầu là 0 rồi cứ cộng phần chênh lệc mỗi chậu với min Height. ko phải lo số âm vì min Height đã luôn đc xác định mới trong for loop
print(energy)
```

Công ty XYZ nay tuyển thêm 1 nhân viên mới. Do đợt vừa rồi có rất nhiều nhân viên nghỉ việc nên có một số mã số nhân viên không sử dụng.

Dòng thứ hai chứa n số nguyên là mã số của các nhân viên. Mọi giá trị đều nằm trong phạm vi cho phép.

Dữ liệu xuất

In ra mã số nhân viên còn trống nhỏ nhất.

Giải thích ví dụ

Công ty có 4 nhân viên với các mã số là 4, 2, 1 và 6.

Hướng dẫn giải

Đọc dữ liệu:

Đọc số n nằm trên một dòng duy nhất.

Đọc dãy gồm các phần tử là số nguyên vào mảng a.

Ta sẽ kiểm tra lần luợt các số 1, 2, 3, ... xem số nào là số đầu tiên không xuất hiện trong mảng a. 

Gọi ans là biến kết quả, ban đầu ans = 1, và bắt đầu lặp: 

- Nếu ans xuất hiện trong mảng a thì ans vẫn chưa phải kết quả cần tìm, do đó ta sẽ tăng ans lên 1 đơn vị và kiểm tra tiếp (ans += 1). 

- Ngược lại, nếu ans không xuất hiện trong mảng thì đây chính là kết quả cần tìm, lúc này ta chỉ cần in ans ra và dừng lặp.

Để kiểm tra một số có xuất hiện trong một mảng hay không: chúng ta sẽ xây dựng một hàm kiểm tra. Hàm này nhận tham số là một số k và mảng arr, nó sẽ kiểm tra xem số k có nằm trong mảng arr hay không và trả về True nếu có, nguợc lại trả về False. Cách cài đặt:

- Xây dựng một hàm với tham số là mảng a, kích thước mảng a (nếu cần), và số k.

- Lặp qua các phần tử của mảng: Nếu phần tử bằng k, hàm trả về true
Nếu lặp xong nhưng vẫn chưa có trả về giá trị thì kết luận k không xuất hiện trong a, trả về false.

```
# Method 1:
n = int(input())  
a = list(map(int, input().split()))  
  
def inArray(k, arr) :  
    for i in range(len(arr)) :  
        if arr[i] == k :  
          return True  
    return False  
  
ans = 1  
while True :  
    if inArray(ans, a):  
        ans += 1  
        continue  # ko cần continue
    else :  
        print(ans)  
        break
```

9/ Trái cây
Hary rất thích ăn trái cây đặc biệt là táo và cam, anh ấy thích táo nhất.

Mẹ của Hary đi chợ về có mua vài giỏ trái cây, mỗi giỏ gồm a quả táo và b quả cam. Hary muốn chọn một giỏ có nhiều táo nhất, nhưng nếu số táo bằng nhau anh ấy sẽ chọn giỏ có nhiều cam nhất trong những giỏ nhiều táo nhất.

Hỏi Hary sẽ chọn giỏ trái cây nào.

Dữ liệu nhập

Dòng đầu tiên chứa số nguyên dương NN là số lượng giỏ trái cây $(N \le 5000)$

N dòng tiếp theo mỗi dòng chứa hai số nguyên không âm $a_i$ và $b_i$ $(a_i, b_i \le 300)$ là số lượng táo và cam của giỏ thứ $i$, các giỏ được đánh số từ 1 đến $N$.

Đảm bảo các giỏ là hoàn toàn khác nhau. Hai giỏ $i$ và $j$ khác nhau nếu $a_i ≠ a_j$ hoặc $b_i​  ≠ b_j$

Dữ liệu xuất

In ra vị trí của giỏ mà Hary chọn.

Ví dụ

3

2 3

1 4

2 5

output

3

Giải thích ví dụ

Hary có 3 giỏ trái cây.

Giỏ thứ nhất có 2 trái táo và 3 trái cam.

Giỏ thứ 2 có 1 trái táo và 4 trái cam.

Giỏ thứ 3 có 2 trái táo và 5 trái cam. Ta thấy giỏ 1 và 3 cùng có nhiều số táo nhất là 2. Nhưng trong 2 giỏ này, giỏ 3 có số cam nhiều hơn giỏ 1. Vậy Hary chọn giỏ 3.

Hướng dẫn giải

Đọc dữ liệu:

Đọc số $n$ nằm trên một dòng duy nhất.
Để đọc được 2 mảng a và b, ban đầu ta khởi tạo a và b rỗng, sau đó đọc từng cặp $a_{i}, b_{i}$
​​  rồi thêm vào a và b.
Chúng ta lại sử dụng kỹ thuật lính canh để giải quyết bài này. Gọi biến lính canh kết quả là $id$ (tức là chỉ số của giỏ trái cây tốt nhất). Ban đầu khởi tạo $id = 0$, tức là tạm coi như giỏ trái cây tốt nhất là giỏ đầu tiên. Sau đó xét lần lượt các giỏ 1, 2, 3, ..., $n - 1$ (bằng vòng lặp for):

Nếu giỏ thứ $i$ "tốt" hơn giỏ thứ id thì ta coi $i$ là giỏ tốt nhất lúc này (tức là cập nhật lại $id = i$). Vậy giỏ thứ i tốt hơn giỏ thứ id khi nào? -> khi $a_{i} > a_{id}$ hoặc $a_{i} == a_{id}$ và $b_{i} > b_{id}$.

Cuối cùng ta in ra giá trị $id + 1$ vì trong đề bài các giỏ trái cây đánh số từ 1 (tức là nếu giỏ tốt nhất là giỏ 0 (trong code) thì in 1, nếu là 1 (trong code) thì in 2, ... nếu là $n - 1$ (trong code) thì in ra $n$).

```
n = int(input())  
a = []  # apple
b = []  # orange 
for i in range(n) :  
    ai, bi = map(int, input().split())  
    a.append(ai)  
    b.append(bi)  
      
id = 0  
for i in range(1, n) :  
    if a[id] < a[i] or (a[id] == a[i] and b[id] < b[i]) :  
        id = i  
      
print(id + 1)
```

10/ Quản lí LALA

Rằng Lala không được phép học Tin học quá 3 tiếng liên tục và sau khi học Tin học xong thì Lala phải dành ra ít nhất một tiếng học Tiếng Anh.

Bạn biết được Lala học môn nào trong mỗi giờ của n giờ. Nhiệm vụ của bạn là kiểm tra xem Lala có thực hiện đúng lời ba mẹ hay không?

Dữ liệu nhập

Dòng đầu tiên chứa số nguyên dương $n$ $(n \le 24)$.

Dòng thứ hai chứa n số nguyên một số 0 hoặc 1. Nếu là 0, giờ thứ i Lala đang học Tin học, ngược lại Lala đang học Tiếng Anh.

Dữ liệu xuất

In ra "YES" nếu Lala thực hiện đúng lời ba mẹ, ngược lại in ra "NO".

Ví dụ

input

5

0 0 0 1 1

output

YES

Giải thích ví dụ

Giờ thứ nhất, giờ thứ hai và giờ thứ ba cô học tin học. Sau đó cô học tiếng Anh vào hai giờ cuối cùng. Như vậy Lala thực hiện đúng lời ba mẹ.

```
# Method 1:
n = int(input())
a = list(map(int, input().split()))
flag = True
time = 0
 
# kiếm tra phần tử cuối nếu là học tin thì cho out luôn
if (a[n - 1] == 0):
    flag = False
    
else:
    for i in range(0, n):
        # nếu LALA học tin
        if (a[i] == 0):
            time = time + 1
            # kiểm tra time > 3?
            if (time > 3): # if này thuộc if trên
                flag = False
        # nếu LALA học tiếng Anh thì time quay về = 0 
        else:
            time = 0 
 
if (flag == True):
    print("YES")
else:
    print("NO")
```

Ý tưởng

Tiếp theo, ta có một nhận xét nhỏ một nhận xét nhỏ:

Đó là nếu giờ học cuối cùng của LaLa là Tin học thì rõ ràng cô đã không học tiếng Anh sau khi học xong giờ Tin này, ta kết luận ngay LaLa không vâng lời rồi in ra NO.

Nếu giờ cuối cùng LaLa học tiếng Anh thì chúng ta sẽ xét xa hơn: bây giờ điều kiện để kết luận LaLa không vâng lời là tồn tại 44 giờ liên tục mà cô học Tin, có nghĩa là tồn tại một vị trí i nào đó mà $a[i] = a[i + 1] = a[i + 2] = a[i + 3]$ và cùng bằng 0.

Việc này gợi ý cho chúng ta sử dụng kỹ thuật đặt cờ hiệu. Ban đầu ta sẽ đặt biến cờ hiệu là $isGood = True$, tức là tin tưởng rằng LaLa vâng lời, sau đó ta sẽ đi kiểm chứng.

Chúng ta duyệt từng vị trí $i = 0, 1, 2, ... n - 4 (i < n - 3)$ vì $i + 3$ phải bé hơn $n$ để việc truy cập $a[i], a[i + 1], a[i + 2] và a[i + 3]$ là đều có nghĩa).

Với mỗi vị trí $i$, nếu mà $a[i] = a[i + 1] = a[i + 2] = a[i + 3] = 0$ thì ta đánh dấu $isGood= False$ và dừng xét, ta kết luận rằng LaLa không vâng lời.

Hướng dẫn giải

Đầu tiên đọc dữ liệu: $n$ và mảng $a$.

Nếu phần tử cuối cùng của $a$ là 0, ta in kết quả là NO.
Ngược lại:

Tạo biến $isGood = true$.

Lặp $i$ từ 0 đến $n - 4$: Nếu $a[i] = a[i + 1] = a[i + 2] = a[i + 3] = 0$, ta gán $isGood = false$

Kiểm tra biến $isGood$ để in ra YES, NO tương ứng.

```
# Method 2:
n = int(input())
a = list(map(int, input().split()))

if a[-1] == 0:
    print('NO')
else:
    isGood = True
    for i in range(n - 3):
        if a[i] == a[i + 1] == a[i + 2] == a[i + 3] == 0: # nếu 0 xuất hiện 4 lần liên tục (quá 3 giờ học Tin) thì False rồi break luôn
            isGood = False
            break
    if isGood:
        print('YES')
    else:
        print('NO')
```
