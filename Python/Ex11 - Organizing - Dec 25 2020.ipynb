1/ Sắp xếp tăng dần

Cho mảng một chiều các số nguyên. Hãy sắp xếp mảng tăng dần.

Mảng được sắp xếp tăng dần là mảng thỏa điều kiện sau:

$A_i \le A_{i+1}$
​​  với $0 \le i < N-1.$

Dữ liệu nhập

Dòng thứ nhất là số phần tử của mảng $N (1 \le N \le 1 000)$.

Dòng thứ hai là NN số nguyên $A_i
​​  (0 \le i < N, |Ai| \le 10^{9})$

Dữ liệu xuất

Một dòng duy nhất gồm N phần tử của mảng sau khi đã sắp xếp lại theo thứ tự tăng dần. Các phần tử cách nhau bởi một khoảng trắng.

Ví dụ

inputcopy

5

6 4 7 5 1

outputcopy

1 4 5 6 7 

Hướng dẫn giải:

Bài này giới hạn số lượng phần tử không vượt quá 1000, nên bạn có thể sử dụng các thuật toán như insertion sort để giải.

Điều kiện giữa hai phần tử $a_i$
​​  và $a_j
​​  (i < j)$ phải được sắp xếp lại là: $a_i < a_j$

Độ phức tạp:$O(n^2)$ với n là số phần tử của mảng.

```
def insertionSort(b, x):
    for i in range(len(b)):
        if (b[i] > x):
            b.insert(i, x)
            return
    b.append(x)
n = int(input()) 
a = list(map(int, input().split()))
b = []
for x in a:
    insertionSort(b, x)
 
for j in range(len(b)):
    print(b[j], end=' ')
```

2/ Sắp xếp giảm dần

Cho mảng một chiều các số nguyên. Hãy sắp xếp mảng giảm dần.

Mảng được sắp xếp giảm dần là mảng thỏa điều kiện sau:

$A_i \ge A_{i+1}$
​​  với $0 \le i < N-1.$

Dữ liệu nhập

Dòng thứ nhất là số phần tử của mảng $N (1 \le N \le 10^{5})$

Dòng thứ hai là NN số nguyên $A_i
​​  (0 \le i < N, |Ai| \le 10^{9})$

Dữ liệu xuất

Một dòng duy nhất gồm NN phần tử của mảng sau khi đã sắp xếp lại theo thứ tự giảm dần. Các phần tử cách nhau bởi một khoảng trắng.

Ví dụ

inputcopy

5

6 4 7 5 1

outputcopy

Hướng dẫn giải:

Bài này giới hạn số lượng phần tử không vượt quá 100 000, nên bạn không thể sử dụng thuật toán Insertion Sort để giải quyết vấn đề. Do đó ta cần sử dụng các thuật toán sắp xếp tốt hơn chẳng hạn như thuật toán Merge Sort.

Tuy nhiên, mình có thể cài đặt Merge Sort theo 2 cách:

C1: Trong hàm Merge, sau khi đã đưa những phần tử từ hai mảng con Left và Right vào một mảng chung aa, có thể vẫn còn phần tử ở Left hoặc Right chưa được đưa vào. Khi đó ta sẽ duyệt tiếp hai vòng lặp ở cả 2 mảng con để đưa các phần tử còn lại đó vào mảng aa.

C2: Dùng kỹ thuật đặt lính canh ở phần tử sau phần tử cuối cùng ở cả 2 mảng con Left và Right. Ý tưởng như sau:

Gọi n1, n2 lần lượt là số lượng phần tử của 2 mảng con Left và Right.

Giả sử toàn bộ mảng Left đã được đưa vào aa nhưng mảng Right vẫn còn phần tử chưa được đưa vào. Khi ấy index của mảng Left sẽ đạt n1, ta đặt lính canh ở đây để khi so sánh ta luôn đẩy được tất cả phần tử còn lại của Right vào.

Lính canh trong bài này sẽ là -1.000.000.001.

Độ phức tạp: O(n log n) với nn là số phần tử của mảng.

```
# Method 1: long time
def insertionSort(b, x):
    for i in range(len(b)):
        if (b[i] < x):
            b.insert(i, x)
            return
    b.append(x)
n = int(input()) 
a = list(map(int, input().split()))
b = []
for x in a:
    insertionSort(b, x)
 
for j in range(len(b)):
    print(b[j], end=' ')
```

```
# Method 3:
inf = -1000000001

def merge(n1, L, n2, R, a):
    L.append(inf)
    R.append(inf)
    i = j = k = 0

    while i < n1 or j < n2:
        if L[i] > R[j]:
            a[k] = L[i]
            i += 1
        else:
            a[k] = R[j]
            j += 1
        k += 1

def mergeSort(n, a):
    if n > 1:
        n1 = n // 2
        n2 = n - n1
        L = a[:n1]
        R = a[n1:]

        mergeSort(n1, L)
        mergeSort(n2, R)
        merge(n1, L, n2, R, a)

n = int(input())
a = list(map(int, input().split()))
mergeSort(n, a)
for ele in a:
    print(ele, end = ' ')
```

4/ Trung vị

Cho số nguyên dương NN và mảng một chiều gồm NN số nguyên $a_i$. Hãy tìm giá trị trung vị của mảng đó.

Biết rằng: Trung vị (Median) trong mảng một chiều là phần tử có giá trị ở giữa mảng. Với trường hợp số lượng phần tử là chẵn thì ta lấy giá trị trung bình của hai phần tử giữa mảng làm Median.

Dữ liệu nhập

Dòng đầu là số nguyên dương $N (1 \le N \le 1000)$.

Dòng tiếp theo gồm N số nguyên $a_i
​​  (0 \le a_i \le 1000)$

Dữ liệu xuất

Một số duy nhất là số trung vị Median.

Ví dụ

inputcopy

5

7 3 1 2 5

outputcopy

3

inputcopy

6

9 3 7 2 1 4

outputcopy

3.5

Giải thích ví dụ

Ví dụ 1:

Mảng sau khi sắp xếp là: 1 2 3 5 7, ta dễ dàng thấy 3 chính là trung vị.

Ví dụ 2:

Mảng sau khi sắp xếp là: 1 2 3 4 7 9, ta có hai phần tử chính giữa là 3 và 4, nên kết quả là trung bình cộng của 3 và 4.

Hướng dẫn giải: Để tìm được trung vị Median theo yêu cầu của đề thì đầu tiên ta phải sắp xếp mảng đã cho lại trước. Bài này số lượng phần tử không vượt quá 1000 nên ta sử dụng Insection Sort hay Merge Sort đều được.

Sau khi sắp xếp mảng xong, ta sẽ có 2 trường hợp:

NN là số lẻ, thì trung vị Median sẽ là phần tử thứ chính giữa của mảng. Tức là, phần tử ở vị trí $\frac{N}{2}$

N là số chẵn, thì trung vị Median sẽ là trung bình cộng của hai phần tử ở giữa mảng. Tức là, trung bình cộng của hai phần tử ở vị trí $\frac{N}{2}$
​​  và $\frac{N}{2} - 1$

Độ phức tạp: $O(N^2)$

 nếu sử dụng thuật toán Insertion Sort, $O(NlogN)$ nếu sử dụng thuật toán Merge Sort.

```
def insectionSort(a, n):
    for i in range(n):
        cur = a[i]
        pos = i
        while pos > 0 and cur < a[pos - 1]:
            a[pos] = a[pos - 1]
            pos -= 1
        a[pos] = cur

n = int(input())
a =  list(map(int, input().split()))
insectionSort(a, n)
if n % 2 != 0:
    print(a[n // 2])
else:
    print((a[n // 2] + a[n // 2 - 1]) / 2)
```

Hướng dẫn giải

Bài này có thể sử dụng thuật toán sắp xếp bình thường, nhưng ta cần lưu ý cách so sánh hai học sinh.

Để dễ dàng, ta nên định nghĩa class học sinh và kèm theo phương thức so sánh 2 học sinh với nhau.

```
class Student:
    def __init__(self, id, point):
        self.id = id
        self.point = point

    def compare(self, other):
        return self.point > other.point or (self.point == other.point and self.id < other.id)

def merge(n1, L, n2, R, a):
    i = j = k = 0
    while i < n1 and j < n2:
        if L[i].compare(R[j]):
            a[k] = L[i]
            i += 1
        else:
            a[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        a[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        a[k] = R[j]
        j += 1
        k += 1

def mergeSort(n, a):
    if n > 1:
        n1 = n // 2
        n2 = n - n1
        L = a[:n1]
        R = a[n1:]

        mergeSort(n1, L)
        mergeSort(n2, R)
        merge(n1, L, n2, R, a)

a = []
n, k = map(int, input().split())
for i in range(n):
    line = input().split()
    a.append(Student(int(line[0]), float(line[1])))

mergeSort(n, a)

for i in range(0, k):
    print(a[i].id)
```
