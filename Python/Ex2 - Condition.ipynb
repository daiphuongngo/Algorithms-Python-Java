1/ Số lớn nhất

Bài toán tìm giá trị lớn nhất của hai số nguyên là một bài toán rất cơ bản trong tin học. Bạn được cho đầu vào 2 số nguyên aa và bb, hãy tìm giá trị lớn nhất của hai số nguyên đó.

Dữ liệu nhập

Một dòng duy nhất chứa hai số nguyên $a, b (|a|, |b| \leq 1000000)$

Dữ liệu xuất

In ra một dòng duy nhất là giá trị lớn nhất của hai số nguyên aa và bb.

Ví dụ

2 3

```
a, b = map(int, input().split())  
if a > b:  
    print(a)  
else:  
    print(b)  
```

2/ Xét dấu

Viết chương trình nhập vào hai số thực aa và bb khác 0. Kiểm tra xem chúng có cùng dấu hay không. Hai số thực được gọi là “cùng dấu” nếu như chúng cùng dương hoặc cùng âm.

Dữ liệu nhập

Một dòng duy nhất chứa hai số thực $a, b khác 0 (|a|, |b| \leq 1000000)$

Dữ liệu xuất

Nếu aa và bb cùng dấu in ra "YES". Ngược lại aa và bb khác dấu in ra "NO".

Ví dụ

2.1 -4.5

```
a, b = map(float, input().split())  
if a * b > 0:  
    print('YES')  
else:  
    print('NO') 
```

4/ Năm nhuận

Upan và Ipan vừa được học về năm nhuận, cô giáo bảo họ rằng, một năm là năm nhuận khi:

Năm đó chia hết cho 4 mà không chia hết cho 100. Ví dụ như 2004, 2012 là các năm nhuận; 1800, 1900, … thì không phải.

Ngoài ra năm nhuận còn có thể là năm chia hết cho 400. Ví dụ như 2000, 1600 là các năm nhuận.

Do trên lớp không nghe giảng kỹ nên hai cậu đã quên mất các quy tắc xác định xem một năm có phải là năm nhuận hay không. Bạn hãy viết một chương trình giúp hai cậu ấy nhé.

Dữ liệu nhập

Ví dụ

1600 --> YES

```
x = int(input())  
if (x % 4 == 0 and x % 100 != 0) or x % 400 == 0:  
    print('YES')  
else:  
    print('NO')  
```

5/ Số may mắn

Upan và Ipan là đôi bạn thân. Hai bạn muốn tìm cho mình chung một con số may mắn. Nhưng với Upan thì một số là số may mắn phải chia hết cho aa. Còn với Ipan một số được gọi là may mắn nếu nó chia hết cho bb.

Biết hai bạn đang cùng nhau tìm một con số may mắn cho riêng mình. Ông bụt hiện ra và trao tặng hai bạn một con số $x$. Số $x$ có thể thỏa mãn cho Ipan hoặc Upan. Hoặc thật tuyệt vời nó khớp với yêu cầu số may mắn của cả hai. Nhưng số $x$ cũng có thể không phải là số may mắn của cả 2 bạn.

Căn cứ vào 3 số $a, b, x$, hãy giúp ông bụt xác định xem ai là người có số may mắn.

Dữ liệu nhập
Dòng đầu tiên chứa hai số nguyên $a và b (1 \leq a, b \leq 100,000)$.

Dòng tiếp theo chứa số nguyên $x (1 \leq x \leq 10,000,000)$.

Dữ liệu xuất
Nếu số $x$ là số may mắn của cả hai in ra “Both”.

Nếu số $x$ chỉ là số may mắn của Upan in ra “Upan”.

Nếu số $x$ chỉ là số may mắn của Ipan in ra “Ipan”.

Nếu số $x$ không là số may mắn của ai trong hai bạn in ra “No”.

Ví dụ

4 9
36


```
a, b = map(int, input().split())  
x = int(input())  
if x % a == 0 and x % b == 0:  
    print('Both')  
elif x % a == 0:  
    print('Upan')  
elif x % b == 0:  
    print('Ipan')  
else:  
    print('No')  
```

7/ Tiền điện

"Ding!!! Dong!!! Ding!!! Dong!!!" - chuông cửa nhà bạn lại vang lên không ngừng vào ngày này mỗi tháng vì nhân viên thu tiền điện lại đến tìm bạn. Bạn được xem hóa đơn tiền điện mình dùng trong tháng trước nhưng băn khoăn không biết rằng con số này có được tính chính xác theo quy định hay không.

Bạn biết được quy định về giá bán điện của Bộ Công Thương như sau:

Bậc	Giá bán lẻ điện sinh hoạt bậc thang	Giá (VND/kWh)

1	Cho kWh từ 0 đến 50	1,484

2	Cho kWh từ 51 đến 100	1,533

3	Cho kWh từ 101 đến 200	1,786

4	Cho kWh từ 201 đến 300	2,242

5	Cho kWh từ 301 đến 400	2,503

6	Cho kWh từ 401 trở lên	2,587

Với chỉ số điện tháng trước và tháng này ghi trên hóa đơn, bạn hãy tự tính lại số tiền điện phải trả để so sánh kết quả.

Lưu ý: tiền phải trả == tiền điện sử dụng + thuế VAT (10% tiền điện). Kết quả chỉ lấy phần nguyên.

Dữ liệu nhập

Gồm một dòng duy nhất chứa hai số nguyên không âm cách nhau bởi một khoảng trắng.

Số đầu tiên là chỉ số điện đầu tháng trước, số tiếp theo là chỉ số điện đầu tháng này của bạn. (Bạn được đảm bảo là chỉ số điện tháng sau luôn không bé hơn chỉ số điện tháng trước và các chỉ số điện có giá trị không quá $10^6$)

Dữ liệu xuất

In ra một số nguyên duy nhất là số tiền bạn phải trả cho lượng điện mình đã sử dụng tháng trước.

Ví dụ

120 168


Giải thích ví dụ

Ví dụ 1:

Tháng trước bạn sử dụng 48 kWh điện (số mới 168 – số cũ 120). Tra theo bảng trên thì bạn chỉ sử dụng điện ở bậc 1 nên tiền điện của bạn là 48 \times 1484 = 7123248×1484=71232 (VND). Vậy tiền phải trả là 71232 + 71232 \times 10\% = 78355.271232+71232×10%=78355.2 (VND), bạn chỉ cần in ra 7835578355.

Ví dụ 2:

Bạn vừa sống tháng đầu tiên ở căn nhà mới mua và sử dụng 203 kWh điện.

50 kWh điện đầu tiên tốn 50 * 1484 = 74200 (VND).

50 kWh điện tiếp theo (từ 51 đến 100) tốn 50 * 1533 = 76650 (VND).

100 kWh điện tiếp theo (từ 101 đến 200) tốn 100 * 1786 = 178600 (VND).

3 kWh điện cuối cùng (từ 201 đến 203) tốn 3 * 2242 = 6726 (VND).

Vậy chi phí điện sinh hoạt của bạn là 74200 + 76650 + 178600 + 6726 = 336176 (VND). 

Và tổng số tiền bạn cần trả là 336176 + 336176 × 10% = 369793.6, vì chỉ lấy phần nguyên nên kết qua cần in ra là 369793.


```
a, b = map(int, input().split())
x = b - a
if x <= 50:
    ans = x * 1484
elif x <= 100:
    ans = 50 * 1484 + (x - 50) * 1533
elif x <= 200:
    ans = 50 * 1484 + 50 * 1533 + (x - 100) * 1786
elif x <= 300:
    ans = 50 * 1484 + 50 * 1533 + 100 * 1786 + (x - 200) * 2242
elif x <= 400:
    ans = 50 * 1484 + 50 * 1533 + 100 * 1786 + 100 * 2242 + (x - 300) * 2503
else:
    ans = 50 * 1484 + 50 * 1533 + 100 * 1786 + 100 * 2242 + 100 * 2503 + (x - 400) * 2587

print(int(ans * 1.1))
```

8/ Số kẹo nhiều nhất

Có 4 túi kẹo được đặt trên bàn và mỗi túi kẹo được dán bên ngoài là số viên kẹo có trong túi. Bé An được phép chọn 1 túi kẹo bất kỳ trong số đó đem về làm quà. Vì kẹo là món ưa thích nên bé An muốn chọn túi có nhiều viên kẹo nhất. Bạn hãy giúp bé An nhé!

Dữ liệu nhập

Gồm một dòng duy nhất lần lượt là số kẹo trong mỗi túi. Biết mỗi túi có không quá 100 viên kẹo, và túi nào cũng có kẹo.

Dữ liệu xuất

In ra số kẹo nhiều nhất mà bé An có thể nhận được.

Ví dụ

8 7 6 10

10

Giải thích ví dụ

Ví dụ 1: Ở ví dụ này, ta thấy:

Túi đầu tiên có 8 viên kẹo.

Túi thứ hai có 7 viên kẹo.

Túi thứ ba có 6 viên kẹo.

Túi thứ tư có 10 viên kẹo.

Vì vậy, số kẹo nhiều nhất mà bé An có thể nhận được là 10 viên kẹo.

```
candy1, candy2, candy3, candy4 = map(int, input().split())

maxCandy = candy1
if (maxCandy < candy2):
    maxCandy = candy2
if (maxCandy < candy3):
    maxCandy = candy3
if (maxCandy < candy4):
    maxCandy = candy4

print(maxCandy)
```
