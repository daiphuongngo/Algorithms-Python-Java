1/ Tính tổng

Viết chương trình đọc vào số nguyên dương nn và tính tổng các số nguyên từ 1 đến nn, tức là (1 + 2 + ... + n).

Dữ liệu nhập

Dòng đầu tiên chứa số nguyên dương $n (n \leq 100)$.

Dữ liệu xuất

In tổng $S$.

```
n = int(input())
res = 0
for i in range(1, n + 1):
    res = res + i
print(res)
```

3/ So sánh

Hôm nay cô giáo phát bài kiểm tra trong lớp. Cô đọc điểm lần lượt từng học sinh.

Sau khi đọc cô muốn bạn thống kê điểm số cao nhất lớp và điểm số thấp nhất lớp.

Bạn hãy viết 1 chương trình nhỏ, để giúp cô giáo nha.

Dữ liệu nhập

Gồm nhiều dòng, mỗi dòng chứa một số nguyên duy nhất là điểm của từng học sinh trong lớp (Số dòng không quá 10000 dòng và luôn ít nhất có điểm của một học sinh, dữ liệu đảm bảo số điểm không nhỏ hơn 0 và không lớn hơn 10). Kết thúc các dòng dữ liệu điểm là một dòng chứa số -1.

Dữ liệu xuất

Một dòng duy nhất chứa hai số nguyên lần lượt là điểm số cao nhất và điểm số thấp nhất cách nhau bởi một khoảng trắng.

Ví dụ

2

7

9

1

-1

```
highest = -1
lowest = 11
while (True) :
    p = int(input())
    if (p == -1) : break
    if p > highest: highest = p;
    if p < lowest: lowest = p;
print(highest, lowest)
```

4/ Số nguyên tố

Số nguyên tố là các số nguyên dương chỉ có hai ước số dương phân biệt là 1 và chính nó. Các số có nhiều hơn 2 ước số dương được gọi là hợp số.

Do số 1 chỉ có duy nhất một ước số dương là chính nó, nên số 1 không phải là số nguyên tố và cũng không phải là hợp số.

Cho một số nguyên dương, kiểm tra số đó có phải số nguyên tố hay không?

Dữ liệu nhập

Gồm 1 dòng duy nhất chứa một số nguyên dương $n (n \le 1000)$.

Dữ liệu xuất

In ra “YES” nếu n là số nguyên tố, ngược lại in ra “NO”.

```
import math

n = int(input())
isPrime = True
if n < 2:
    isPrime = False
for i in range(2, int(math.sqrt(n)) + 1):
    if (n % i == 0):
        isPrime = False
        break
if (isPrime):
    print("YES")
else:
    print("NO")
```

5/ Dãy tăng

Lulu có một vườn hoa có n chậu. Mỗi chậu hoa có một độ cao nhất định.

Tết đến rồi Lulu mong muốn có một vườn hoa mặt trời như ý. Một vườn hoa mà Lulu cảm thấy thích là khi các chậu hoa phía sau chắc chắn phải cao hơn hoặc bằng chậu hoa ở phía trước.

Lulu thuê những người thợ làm vườn tới để tỉa chậu hoa theo ý của anh ấy. Nhưng hôm nay anh ấy có việc phải ra ngoài, anh ấy nhờ bạn kiểm tra, liệu với độ cao những chậu hoa sau khi tỉa thì có phù hợp với yêu cầu của anh ấy hay không?

Dữ liệu nhập

Gồm nhiều dòng, mỗi dòng là một số nguyên dương là chiều cao của chậu hoa tương ứng.

Kết thúc bộ dữ liệu là một dòng chỉ chứa số 0. Số dòng của bộ dữ liệu không vượt quá 10,000 và chiều cao của các chậu hoa luôn lớn hơn 0 và nhỏ hơn hay bằng 1,000,000.

Dữ liệu xuất

In ra “YES” nếu các chậu cây sau khi tỉa là thỏa mãn yêu cầu của Lulu. Ngược lại in ra “NO”.

```
res = True
pre = 0
while (True) :
    cur = int(input())
    if (cur == 0): break
    if (cur < pre):
        res = False
        break
    pre = cur # để kiểm tra các số sau
if (res):
    print("YES")
else:
    print("NO")
```

6/ Chia kẹo

Mẹ của Lu và Xi vừa mua n túi kẹo, mỗi túi có một số lượng kẹo nhất định.

Mẹ muốn chia kẹo cho hai anh em, nhưng mẹ muốn với mỗi bịch đều có thể chia đều cho cả hai để không gây mâu thuẫn. Hỏi mẹ của Lu và Xi có thể nào thực hiện được điều đó không.

Dữ liệu nhập

Dòng đầu tiên là số nguyên dương $n (1 \le n \le 1000)$.

n dòng tiếp theo mỗi dòng chứa một số nguyên $x (1 \le x \le 1000)$ là số lượng kẹo của mỗi bịch.

Dữ liệu xuất

In ra “YES” nếu kẹo mỗi bịch đều có thể chia đều cho hai anh em Lu và Xi.

Ngược lại in ra “NO”

```
n = int(input())
flag = True
for i in range(0, n):
    x = int(input())
    if (x % 2 != 0):
        flag = False
        break
if (flag):
    print("YES")
else:
    print("NO")
```

7/ Hàng rào

Viết chương trình vẽ hàng rào nhà Harry có dạng hình vuông có độ dài n. Hàng rào được kí hiệu bởi kí tự '*', các ô đất bên trong hàng rào là các khoảng trắng.

Dữ liệu nhập

Một dòng duy nhất chứa một số nguyên dương $n (2 \le n \le 100)$.

Dữ liệu xuất

Hình dạng hàng rào nhà Harry.

```
n = int(input())

for i in range(1, n + 1):
    print('*', end = '')
print()

for i in range(2, n) :
    print('*', end = '')
    for j in range(2, n) :
        print(' ', end = '')
    print('*')

for i in range(1, n + 1) :
    print('*', end = '')
```

Ta thấy rằng với một hàng rào hình vuông có độ dài cạnh là n, thì hàng 1 và hàng n sẽ gồm n dấu * được viết liên tục. Trong khi đó trên mỗi hàng từ hàng 2 đến hàng n – 1, chỉ gồm một dấu * ở vị trí đầu và một dấu * ở vị trí cuối, còn lại giữa 2 ký tự * đó hoàn toàn không có gì.

Như vậy, ta sẽ in ra được hàng rào bằng cách như sau:

Ta in ra hàng rào thứ 1, bằng cách duyệt một vòng lặp lặp lại n lần, mỗi lần in ra dấu *, và phải nhớ chuyển xuống dòng dưới để in tiếp.

Ta lặp lại n – 2 lần, tương ứng với việc vẽ hàng rào trên các hàng từ 2 đến n – 1, với mỗi lần lặp, ta in ra dấu * đầu tiên. Tiếp theo là ta lại lặp lại n – 2 lần và in tiếp ra màn hình ký tự khoảng trắng. Cuối cùng ta in ra thêm một dấu * và chuyển tiếp xuống dòng dưới.

Cuối cùng, ta duyệt một vòng lặp lặp n lần, mỗi lần in ra dấu * nhằm mục đích in ra hàng rào ở hàng cuối cùng.

Cách khác: Ta nhân thấy những ô nằm trên hàng rào sẽ thỏa 1 trong 4 điều kiện sau:

i = 0 hoặc i = n - 1 hoặc j = 0 hoặc j = n - 1

Vì thế ta có thể làm như sau:

Lặp i từ 0 đến n - 1:

Lặp j từ 0 đến n - 1:

Nếu i == 0 hoặc i == n - 1 hoặc j == 0 hoặc j == n - 1: - In "*"

Ngược lại: - In " "

Nếu j == n - 1: - In xuống hàng.

Lưu ý: hàm print trong Python mặc định sau khi in sẽ xuống dòng, Nên nếu ta muốn thay đổi sau khi in vẫn giữ trên cùng một hàng thì ta khai báo end = ‘’.
