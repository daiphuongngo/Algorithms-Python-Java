1/ Tính giai thừa

Viết chương trình tính giai thừa của một số nn (kí hiệu là n!) bằng phương pháp đệ quy, biết:

T(n) = n! = 1 x 2 x 3 x … x n

Biết rằng: T(0) = 1.

Dữ liệu nhập

Một dòng duy nhất gồm một số nguyên không âm nn (n $\le$ 10).

Dữ liệu xuất

Một dòng duy nhất là giá trị của T(n).

Ví dụ

inputcopy

3

outputcopy

6

Hướng dẫn giải:

Recursive case:

T(n) = n * T(n – 1).

Base case:

T(0) = 1.

Ví dụ: n = 3, tính T(3).

Dựa vào recursive case:

T(3) = 3 * T(2)).

T(2) = 2 * T(1).

T(1) = 1 * T(0).

T(0) = 1.

Dừng.

Các bước thực hiện:

Đọc vào số nguyên n.

Hàm đệ quy factorial(n):

Nếu n = 0: Trả về 1.

(Ngược lại) Trả về n * factorial(n – 1)!.

In ra kết quả của factorial(n).

Độ phức tạp: O(n).

```python
def factorial(n):
    if n == 1 or n==0:
        return 1
    else:
        return n * factorial(n-1)

x = int(input())
print(factorial(x))
```

2/ Ước lẻ lớn nhất

Viết chương trình tìm ước số lẻ lớn nhất của của một số nguyên dương n bằng phương pháp đệ quy.

Dữ liệu nhập

Một dòng duy nhất là số nguyên dương nn (n $\le$ 10^6)

Dữ liệu xuất

Một dòng duy nhất là ước số lẻ lớn nhất của n.

Ví dụ

inputcopy

100

outputcopy

25

Giải thích ví dụ

Các ước số của 100 là: 1 2 4 5 10 20 25 50 100, nên kết quả là 25.

Hướng dẫn giải:

Gọi hàm f(n, x)f(n,x) là hàm trả về ước lẻ lớn nhất của nn khi đang thử xx.

• Recursive case:

f(n, x) = f(n, $\frac{x}{2}$), nếu x là chẵn, thì ta sẽ thử $\frac{x}{2}$

• Base case:

f(n, x) = x là số lẻ và n chia hết cho x.

Gọi hàm:

Nếu n lẻ, ước lẻ lớn nhất của nn chính là n.

Ngược lại, f(n, $\frac{n}{2}$)

Ví dụ: n = 12. Tìm ước số lẻ lớn nhất của n, ta bắt đầu với f(12, 12).

f(12, 12) gọi f(12, 6) vì x = 12 không phải là số lẻ.

f(12, 6) gọi f(12, 3) vì x = 6 không phải là số lẻ.

f(12,3) dừng đệ quy trả về 3 vì x = 3 là số lẻ và 12 chia hết cho 3.

Các bước thực hiện:

Thiết kế hàm đệ quy maxOddDivisor(n, x):

Nếu x lẻ và nn chia hết cho x: Trả về x. (Ngược lại) Trả về maxOddDivisor(n, $\frac{x}{2}$)

Đọc vào số nguyên n.

In ra maxOddDivisor(n, n).

Độ phức tạp: O(log(n)).

```python
def max_odd_divisor(n, x):
    if x % 2 == 1 and n % x == 0:
        return int(x)
    return max_odd_divisor(n, x // 2)


n = int(input())
print(max_odd_divisor(n, n))
```

3/ Đếm số lượng chữ số

Nhập vào một số nguyên N.

Bạn hãy viết chương trình đệ quy để đếm số lượng chữ số của số nguyên N.

Dữ liệu nhập

Một dòng duy nhất là số nguyên NN (|N| $\le$ 10^6)

Dữ liệu xuất

Một dòng duy nhất là số lượng chữ số của số nguyên N.

Ví dụ

inputcopy

345234

outputcopy

6

Hướng dẫn giải:

Gọi hàm c(n) là hàm đếm số lượng chữ số của nn, ta sẽ tính bằng cách bỏ lần lượt từng chữ số của n ra.

Recursive case:

c(n) = 1 + c($\frac{n}{10}$) nếu n $\ge$ 10.

Base case:

c(n) = 1 nếu n < 10.

Ví dụ: n = 2018. Tìm số lượng chữ số của n.

c(2018) = 1 + c(201).

c(201) = 1 + c(20).

c(20) = 1 + c(2).

c(2) = 1c(2)=1.

Dừng.

Các bước thực hiện:

Thiết kế hàm đệ quy cntDigits(n):

Nếu n < 10: Trả về 1.

(Ngược lại) Trả về 1 + cntDigits($\frac{n}{10}$)

Đọc vào số nguyên n.

Nếu n âm: In ra cntDigits(–n).

Ngược lại: In ra cntDigits(n).

Độ phức tạp: O(|n|) với |n|là số chữ số của n.

```python
import sys
sys.setrecursionlimit(10**9)

def cntDigits(n):
    if (n < 10):
        return 1
    return 1 + cntDigits(n // 10)

n = int(input())
if n < 0:
    print(cntDigits(-n))
else:
    print(cntDigits(n))
```

Dữ liệu nhập

Một dòng duy nhất là số nguyên N |N| $\le$ 10^6).

Dữ liệu xuất

Một dòng duy nhất là chữ số đầu tiên của số N.

Ví dụ

inputcopy

-123

Hướng dẫn giải:

Recursive case:

c(n) = firstDigit(n / 10) nếu n $\ge$ 10

Base case:

c(n) = n nếu n < 10

Ví dụ: n = -123. Tìm chữ số đầu tiên của n.

c(123) = firstDigit(12)

c(12) = firstDigit(1)

Các bước thực hiện:

Thiết kế hàm đệ quy firstDigit(n):

Nếu n < 10: Trả về n

(Ngược lại) Trả về firstDigit(n / 10)

Đọc vào số nguyên n.

Nếu n âm thì đổi dấu của n

Độ phức tạp: O(|n|) với |n| là số chữ số của n.

```python
import sys
sys.setrecursionlimit(10**9)
 
def firstDigit(n):
    if n < 10:
        return n
    return firstDigit(n // 10)
 
n = int(input())
if n < 10:
	n = -n
print(firstDigit(n))
```

5/ Tìm chữ số lớn nhất

Viết chương trình tìm chữ số lớn nhất của số nguyên NN bằng phương pháp đệ quy.

Dữ liệu nhập

Một dòng duy nhất là số nguyên N (|N| $\le$ 10^6).

Dữ liệu xuất

Một dòng duy nhất là chữ số lớn nhất của số nguyên N.

Ví dụ

inputcopy

4252

outputcopy

5

Hướng dẫn giải:

Recursive case:

c(n) = max(n \% 10, c(n // 10)) nếu n $\ge$ 10

Base case:

c(n) = nc(n)=n nếu n < 10

Ví dụ: n = 4252. Tìm số lượng chữ số của n.

c(4252) = max(2, c(425))

c(425) = max(5, c(42))

c(42) = max(2, c(4))

c(4) = 4

Dừng.

Các bước thực hiện:

Thiết kế hàm đệ quy maxDigit(n):

Nếu n < 10: Trả về n

(Ngược lại) Trả về max(n % 10, maxDigit(n // 10))

Đọc vào số nguyên n.

Nếu n âm:

In ra maxDigit(maxDigit(–n))

Ngược lại:

In ra maxDigit(n)

Độ phức tạp: O(|n|) với |n| là số chữ số của n.

```python
import sys
sys.setrecursionlimit(10**9)

def maxDigit(n):
    if n < 10:
        return n
    return max(n % 10, maxDigit(n // 10))

n = int(input())

if n < 0:
    print(maxDigit(-n))
else:
    print(maxDigit(n))
```

6/ Tổng chẵn

Viết chương trình tính tổng các phần tử là số chẵn trong mảng 1 chiều số nguyên bằng phương pháp đệ quy.

Dữ liệu nhập

Dòng đầu tiên chứa số nguyên dương NN là số lượng phần tử trong mảng (N \le 10 000).

Dòng thứ hai là N số nguyên $a_i$ (|$a_i$| $\le$ 10000, 0 $\le$ i $<$ N)

Dữ liệu xuất

Một dòng duy nhất là tổng các phần tử là số chẵn trong mảng.

Ví dụ

inputcopy

3

4 2 5

outputcopy

6

```python
import sys
sys.setrecursionlimit(10**6)

def sumEven(a, n):
    if n == 0:
        return 0
    if a[n - 1] % 2 == 0:
        return a[n - 1] + sumEven(a, n - 1)
    return sumEven(a, n - 1)

n = int(input())
a = list(map(int, input().split()))
print(sumEven(a, n))
```

8/ Số nhị phân

Cho số nn ở hệ thập phân, hãy viết nn dưới dạng số nhị phân.

Cách đổi: Nguyên tắc của phương pháp này là lấy số cần chuyển đổi chia cho 2 (kết quả chỉ lấy phần nguyên), sau đó tiếp tục lấy kết quả chia 2 (và cũng chỉ lấy phần nguyên), kết quả số nhị phân thu được là tập hợp các số dư của các phép chia.

Ví dụ:

n = 10

10 / 2 = 5 dư 0

 5 / 2 = 2 dư 1

 2 / 2 = 1 dư 0

 1 / 2 = 0 dư 1

Kết quả là: 1010

Dữ liệu nhập

Một dòng duy nhất là số nguyên không âm n (n $\le$ $10^{9}$)

Dữ liệu xuất

Một dòng duy nhất là biểu diễn số nn ở hệ nhị phân.

Ví dụ

inputcopy

10

outputcopy

1010

Hướng dẫn giải

Base case: n == 0, in ra 0

Recursive case:

Nếu n // 2 > 0, gọi đệ qui với tham số là n // 2.

In ra n % 2

Độ phức tạp: Với n là tham số đệ qui, ta sẽ cần $\lceil \log_2(n)\rceil$

​​ (n)⌉ bước gọi đệ qui để tìm ra dạng nhị phân của n. Độ phức tạp $O(\lceil \log_2(n)$

```python
import sys
sys.setrecursionlimit(10**6)

def f(n):
  if n == 0:
    print(0, end='')
    return
  if n // 2 > 0:
    f(n // 2)
  print(n % 2, end='')
  
n = int(input())
f(n)
```

9/ Số thập lục phân

Cho số n ở hệ thập phân, hãy viết n dưới dạng số thập lục phân.

Cách đổi cũng như số nhị phân nhưng các giá trị số dư từ 10 trở lên sẽ thành chữ cái tương ứng bắt đầu từ 'A': 10 ('A'), 11 ('B'), 12 ('C'), 13 ('D'), 14 ('E'), 15 ('F').

Ví dụ:

26 = 1 * $16^{1}$ + 10 * $16^{0}​​$ , nên 26 được viết thành $1A$

Dữ liệu nhập

Một dòng duy nhất là số nguyên không âm n (n $\le 10^{9}$).

Dữ liệu xuất

Một dòng duy nhất là biểu diễn số nn ở hệ thập lục phân.

Ví dụ

inputcopy

10

outputcopy

A

Hướng dẫn giải

Base case: n == 0: in ra 0.

Recursive case:

Nếu n / 16 lớn hơn 0: gọi đệ quy với tham số n // 16

In ra kí tự biểu diễn n % 16 trong hệ thập lục phân.

Độ phức tạp: Với nn là tham số đệ qui, ta sẽ cần $\lceil \log_{16}(n)\rceil = \lceil\frac{ \log_2(n)}{4}\rceil$ bước gọi đệ qui để tìm ra dạng thập lục phân của nn. Độ phức tạp $O(\log_2(n))$

```python
import sys
sys.setrecursionlimit(10**6)

def convert2Hex(n):
	if n < 10:
		return n
	return chr(ord('A') + n - 10)

def f(n):
	if n == 0:
		print(0, end='')
		return
	if n // 16 > 0:
		f(n // 16)
	print(convert2Hex(n % 16), end='')
  
n = int(input())
f(n)
```

10/ Chuỗi đối xứng

Cho một chuỗi s có độ dài nn bao gồm các kí tự in hoa, thường và chữ số. Hãy kiểm tra xem chuỗi ss có phải là chuỗi đối xứng không.

Chuỗi đối xứng là chuỗi khi đọc từ trái sang phải hay phải sang trái đều giống nhau.

Dữ liệu nhập

Dòng đầu tiên là số nguyên dương n (n $\le 10^{5}$)

Dòng thứ hai là chuỗi s

Dữ liệu xuất

Nếu s là chuỗi đối xứng, in "YES", ngược lại in "NO".

Ví dụ

inputcopy

5

aBcBa

outputcopy

YES

inputcopy

3

abb

outputcopy

NO

Hướng dẫn giải

Sử dụng đệ qui với ý tưởng kiểm tra tuần tự cặp kí tự ii và n - i - 1.

Base case: i >= j: thì trả về true/True

Ngược lại:

nếu kí tự ii bằng kí tự j (j = n - i - 1): kiểm tra tiếp kí tự i + 1 và j - 1.

Ngược lại trả về false/False

Độ phức tạp: Với s là chuỗi đầu vào, gọi n = |s| là số characters của s. Ta sẽ cần $\lceil \frac{n}{2} \rceil$ bước gọi đệ qui để kiểm tra tính đối xứng của s. Độ phức tạp $O(\lceil \frac{n}{2} \rceil)$

```python
import sys
sys.setrecursionlimit(10**6)

def isSymmetric(s, i, j):
  if i >= j:
    return True
  if s[i] == s[j]:
    return isSymmetric(s, i + 1, j - 1)
  return False
  
  
n = int(input())
s = input()
if (isSymmetric(s, 0, n - 1)):
  print('YES')
else:
  print('NO')
```
