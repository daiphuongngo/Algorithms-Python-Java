1/ Kiểm tra ký tự xuất hiện

Cho một chuỗi kí tự, hãy kiểm tra xem có xuất hiện một trong các ký tự ‘B’, ‘I, ‘G’ hoặc ‘O’ trong chuỗi hay không? (Không phân biệt chữ hoa hay chữ thường). Nếu có in ra "YES", ngược lại in ra "NO".

Dữ liệu nhập

Một dòng duy nhất chứa chuỗi kí tự có độ dài không quá 1000 ký tự.

Dữ liệu xuất

In ra “YES” hoặc “NO” tương ứng với điều kiện như mô tả (không có dấu ngoặc kép).

Hướng dẫn giải:

Bài toán yêu cầu chúng ta kiểm tra từng ký tự có thuộc một trong các ký tự ‘B’, ‘I, ‘G’ hoặc ‘O’. Do không phân biệt hoa hay thường nên các ký tự có thể là 'b', 'i', 'g', 'o'.

Vậy mình tạo một cái mảng chứa các ký tự 'B', 'b', 'I', 'i', 'G', 'g', 'O', 'o'.

Ta sẽ duyệt từng ký tự kiểm tra xem ký tự đó có thuộc một trong các ký tự của mảng đã tạo hay không. Và ta sẽ dùng một biến flag để đánh dấu, nếu trong quá trình duyệt nếu có một ký tự thuộc mảng trên thì flag = True. Sau khi duyệt xong thì ta sẽ dựa vào flag để in ra "YES" hay "NO".

Độ phức tạp: O(|S|)O(∣S∣) - với SS là độ dài của chuỗi.

```python
ls = ['B', 'b', 'I', 'i', 'G', 'g', 'O', 'o']
mystr = input()
flag = False
for s in mystr:
	if s in ls:
		flag = True
		break
if flag:
	print("YES")
else:
	print("NO")
```

2/ Kiểm tra ký tự số

Viết chương trình đếm xem chuỗi đã cho có bao nhiêu ký tự là số.

Dữ liệu nhập

Một dòng duy nhất là chuỗi ban đầu, bao gồm các ký tự từ ‘A’ → ‘Z’, ‘a’ → ‘z’, ’0’ → ‘9’ và khoảng trắng, độ dài của chuỗi không quá 1000 ký tự.

Dữ liệu xuất

Một dòng duy nhất là số lượng ký tự là số trong chuỗi.

Hướng dẫn giải:

Dùng một biến countNum để đếm số lượng ký tự là số, lúc đầu countNum = 0

Duyệt từng ký tự trong chuỗi, nếu mỗi ký tự thỏa điều kiện '0' $\le$ s $\le$ '9
​​  thì tăng biến countNum lên một.

Sau khi duyệt xong thì in biến countNum ra.

Độ phức tạp: O(|S|) - với SS là độ dài chuỗi S.

```python
s = input()

countNum = 0
for ch in s:
    if '0' <= ch <= '9':
        countNum += 1

print(countNum)
```

3/ Chuẩn hóa chuỗi

Một chuỗi được xem là đã được chuẩn hóa khi các từ trong chuỗi cách nhau đúng một khoảng trắng và không có khoảng trắng ở đầu hoặc cuối chuỗi. Bạn hãy viết một chương trình chuẩn hóa chuỗi.

Dữ liệu nhập

Một dòng duy nhất là chuỗi ban đầu, bao gồm các ký tự từ ‘A’ đến ‘Z’, ‘a’ đến ‘z’ và khoảng trắng, độ dài của chuỗi không quá 1.000 ký tự.

Dữ liệu xuất

Một dòng duy nhất là chuỗi sau khi được chuẩn hóa.

Hướng dẫn giải:

Ta dùng biến res để lưu chuỗi sau khi mà mình đã chuyển đổi thành câu, ban đầu res = "".

Một biến Boolean needASpace để đánh dấu khi nào thì mình cần thêm khoảng trắng. Ban đầu biến needASpace = False.

Duyệt qua từng ký tự trong chuỗi. Với mỗi ký tự ta sẽ kiểm tra nếu:

Ký tự hiện tại là khoảng trắng và độ dài chuỗi res lớn hơn 0 (vì ta không muốn thêm khoảng trắng ở đầu câu) thì ta cập nhật needASpace = True.

Ký tự hiện tại không là khoảng trắng thì ta sẽ xét xem nếu needASpace là True, thì ta sẽ thêm khoảng trắng vào, sau đó mới thêm ký tự hiện tại vào res đồng thời gán needASpace = False, ngược lại nếu needASpace là False thì ta thêm ký tự hiện tại vào res.

Biến res cuối cùng sẽ là chuỗi mà mình mong muốn.

Độ phức tạp: O(n)O(n) với nn là độ dài của chuỗi.

```python
str = input()

res = ""
resLen = 0
needASpace = False
for ch in str:
    if ch == ' ':
	    if resLen > 0:
		    needASpace = True
    else:
        if needASpace:
            res += ' '
            resLen += 1
            res += ch
            resLen += 1
            needASpace = False
        else:
            res += ch
            resLen += 1

print(res)
```

4/ Chuẩn hóa tên

Trước khi giải Ngoại hạng anh (Premier league) diễn ra, thì ban tổ chức nhận được danh sách NN cầu thủ được đăng ký thi đấu nhưng họ tên của các cầu thủ chưa được chuẩn hóa. Vì vậy ban tổ chức nhờ bạn viết một chương trình chuẩn hóa lại họ tên của cầu thủ sao cho viết hoa ký tự đầu của mỗi từ và viết thường các ký tự còn lại.

Dữ liệu nhập

Dòng đầu tiên là số nguyên dương NN (N \leq 100)(N≤100) với NN là số cầu thủ được đăng ký thi đấu.

NN dòng tiếp theo, mỗi dòng là một chuỗi ký tự họ và tên các từ cách nhau bằng một khoảng trắng, bao gồm các ký tự từ ‘A’ đến ‘Z’, ‘a’ đến ‘z’ và khoảng trắng, độ dài của chuỗi không quá 100 ký tự.

Dữ liệu xuất

NN dòng, mỗi dòng là chuỗi họ và tên sau khi đã chuyển theo đúng định dạng mong muốn.

Ví dụ

inputcopy

3

WiLiaN borGes Da SiLva

kevin de bruyne

pauL Pogba

outputcopy

Wilian Borges Da Silva

Kevin De Bruyne

Paul Pogba

Hướng dẫn giải:

Đọc N dòng, mỗi dòng ta chuyển thành theo đúng chuẩn Họ và Tên.

Việc chuyển đổi chuỗi theo đúng định dạng như sau:

Duyệt từng ký tự trong chuỗi:

Nếu ký tự ở đầu câu hoặc phía trước nó là một khoảng trắng thì ta viết hoa ký tự này lên (do đây là bắt đầu của một từ mới).

Ngược lại thì ta sẽ viết thường.

Độ phức tạp: O(N * length(S)) với N là số cầu thủ được đăng lý thi đấu, length(S) là độ dài của họ tên cầu thủ.

```python
n = int(input())
for j in range(n) :
    name = input()

    res = ""
    for i in range(len(name)) :
        if i == 0 or name[i - 1] == ' ' :
            if 'a' <= name[i] <= 'z' :
                res += chr(ord(name[i]) - 32)
            else :
                res += name[i]
        else:
            if 'A' <= name[i] <= 'Z' :
                res += chr(ord(name[i]) + 32)
            else :
                res += name[i]

    print(res)
```

5/ Ký tự xuất hiện ít nhất

Cho N chuỗi, hãy viết chương trình tìm ký tự xuất hiện ít nhất trong các chuỗi, chuỗi ký tự chỉ bao gồm các kí tự từ ‘A’ → ‘Z’ và ‘a’ → ‘z’.

Lưu ý: không phân biệt chữ hoa hay chữ thường.

Dữ liệu nhập

Dòng đầu tiên là số lượng chuỗi NN (1 $\leq$ N $\leq$ 1.000).

NN dòng tiếp theo, mỗi dòng là một chuỗi ký tự có độ dài không quá 1.000 ký tự.

Dữ liệu xuất

N dòng, mỗi dòng là một ký tự in hoa của ký tự xuất hiện ít nhất trong chuỗi, nếu nhiều ký tự xuất hiện ít nhất thì xuất ra ký tự có thứ tự từ điển nhỏ nhất.

Lưu ý: Thứ tự từ điển theo chiều tăng dần là từ ‘A’ đến ‘Z’.

Ví dụ

3

cd

A

Hướng dẫn giải:

Các bước chính

Đọc N chuỗi.

Ta viết một hàm nhận một chuỗi và trả lại ký tự xuất hiện ít nhất.
Do không phân biệt chữ hoa hay thường nên để dễ dàng ta sẽ chuẩn hóa chuỗi thành chữ hoa hết (hoặc chữ thường), thế nên sẽ có 26 ký tự A – Z, vì vậy ta khởi tạo một mảng countcount có 26 phần từ dùng để đếm số lần xuất hiện của từng ký tự.

VD: với index = 0 thì tương ứng với A (a). với index = 1 thì tương ứng với B (b).

Vậy làm sao từ kí tự 'a', 'A' ta ra được index = 0, hay kí tự 'b', 'B' ta ra được index = 1?

Ta sẽ viết một hàm để lấy vị trí của một ký tự tương ứng trong mảng. Ta có thể thực hiện được bằng cách lấy mã ASCII của ký tự hiện tại trừ đi mã ASCII của ký tự mốc là 'A' (hoặc 'a' nếu là chữ thường) kết quả có được chính là vị trí của ký tự này trong mảng countcount.

Lưu ý: Với python ta có thể sử dụng dictionary.

Độ phức tạp: O(N * length(S)) với N là số chuỗi, length(S) là độ dài các chuỗi.

```python
def pos(ch) :  
    if 'A' <= ch <= 'Z' :
        return ord(ch) - ord('A')
    return ord(ch) - ord('a')  
  
def getMinimumOccurringChar(str) :
    count = [0] * 26
    for ch in str :
        count[pos(ch)] += 1
    min = 1001
    res = ""
    for i in range(len(count)) :
        if (count[i] != 0 and count[i] < min) :
            min = count[i]
            res = chr(i + ord('A'))
    return res
  
n = int(input())  
for i in range(n) :  
    str = input()  
    print(getMinimumOccurringChar(str))
```

6/ Đếm kí tự phân biệt

Viết chương trình đếm số lượng ký tự trong một chuỗi.

Lưu ý: phân biệt chữ hoa và chữ thường.

Dữ liệu nhập

Một dòng duy nhất là chuỗi ban đầu, bao gồm các ký tự từ 'A' → 'Z' và 'a' → 'z', độ dài của chuỗi không quá 1.000 ký tự.

Dữ liệu xuất

Một dòng duy nhất là số lượng ký tự khác nhau.

Ví dụ

inputcopy

abccccbbb

outputcopy

3

Hướng dẫn giải:

Ta sẽ dùng mảng có 52 phần tử để lưu xem kí tự thứ i đã xuất hiện chưa.
Mảng sẽ lưu các kí tự từ ‘A’ … ‘Z’, ‘a’ … ‘z’.


Những kí tự in hoa ta sẽ đánh số từ 0 đến 25 theo thứ tự từ A đến Z.

Những kí tự in thường ta sẽ đánh số từ 26 đến 51 theo thứ tự từ a đến z.

Sau khi duyệt các kí tự trong chuỗi và đếm số lần xuất hiện của một ký tự, ta sẽ duyệt lại 52 phần tử này để đếm xem bao nhiêu phần tử có số lần xuất hiện khác không.

Mã giả:

Khởi tạo mảng đếm có 52 phần tử, gán tất cả giá trị bằng 0.

Có một hàm trả về vị trí của một ký tự trong mảng đếm.

Duyệt qua từng kí tự trong chuỗi và tăng số lần xuất hiện của ký tự đó trên mảng count.

Duyệt mảng count, những phần tử có số lần xuất hiện khác không thì ta tăng biến đếm số tự phân biệt lên 1.

Lưu ý: Với python ta có thể sử dụng dictionary.

Độ phức tạp: O(N)O(N) với NN độ dài chuỗi.

```python
def pos(ch):
	if 'A' <= ch <= 'Z': return ord(ch) - ord('A')
	return ord(ch) - ord('a') + 26

str = input()
count = [0] * 52
for ch in str: count[pos(ch)] += 1

cnt = 0
for e in count:
	if e != 0: cnt += 1
print(cnt)
```

7/ In hoa sau dấu chấm

Cho một chuỗi ký tự, hãy viết chương trình in hoa các ký tự đứng sau lần cặp kí tự dấu chấm và khoảng trắng ( '. ' : dấu chấm đứng trước, khoảng trắng đứng sau).

Dữ liệu nhập

Một dòng duy nhất là chuỗi ban đầu, bao gồm các ký tự từ 'A' → 'Z', 'a' → 'z', khoảng trắng và dấu chấm, độ dài của chuỗi không quá 1.000 ký tự.

Dữ liệu xuất

Một dòng duy nhất là chuỗi sau khi được in hoa theo cầu.

Ví dụ

inputcopy

I love you. more. than i can. say

outputcopy

I love you. More. Than i can. Say

```python
s = input()
ans = ''
for i in range(len(s)):
    if i >= 2 and s[i - 2] == '.' and s[i - 1] == ' ' and s[i] >= 'a' and s[i] <= 'z':
        ans += chr(ord(s[i]) - ord('a') + ord('A'))
    else:
        ans += s[i]
print(ans)
```
